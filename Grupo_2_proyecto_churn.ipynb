# 1. Introducción

La retención de clientes constituye uno de los principales desafíos para las empresas del sector de las telecomunicaciones. En un mercado altamente competitivo, la pérdida de clientes (*customer churn*) impacta directamente en los ingresos, incrementa los costos asociados a la adquisición de nuevos usuarios y reduce el valor de vida del cliente (*Customer Lifetime Value*). En este contexto, la capacidad de identificar tempranamente a los clientes con mayor probabilidad de abandonar un servicio permite a las organizaciones diseñar estrategias de retención más efectivas, optimizar la asignación de recursos y mejorar la rentabilidad del negocio **(Imani et al., 2025)**.

Desde la perspectiva empresarial, la predicción de *customer churn* permite anticipar qué clientes presentan una mayor probabilidad de abandonar la compañía, facilitando la implementación de campañas de retención focalizadas y una utilización más eficiente de los recursos comerciales. En lugar de aplicar estrategias de fidelización de forma masiva, las organizaciones pueden concentrar sus esfuerzos en aquellos clientes con mayor riesgo de abandono, incrementando la probabilidad de retención y reduciendo los costos asociados a la adquisición de nuevos clientes. Por esta razón, la predicción de *customer churn* se ha convertido en una de las aplicaciones más relevantes del aprendizaje supervisado en sectores como las telecomunicaciones, la banca, los seguros y los servicios financieros **(Imani et al., 2025)**.

El aprendizaje supervisado se ha consolidado como una de las principales herramientas para abordar este problema, permitiendo construir modelos predictivos a partir de datos históricos etiquetados con el objetivo de estimar la probabilidad de abandono de un cliente antes de que éste ocurra. La literatura reciente evidencia un crecimiento sostenido en la aplicación de técnicas de aprendizaje automático para la predicción de *customer churn*, destacando el uso de algoritmos como Regresión Logística, Análisis Discriminante, Naive Bayes, *k-Nearest Neighbors* (k-NN), Árboles de Decisión, Random Forest y *Support Vector Machines* (SVM), los cuales han demostrado un desempeño competitivo dependiendo de las características del conjunto de datos y de las métricas utilizadas para evaluar su rendimiento **(Imani et al., 2025)**.

Desde el punto de vista científico, la predicción de *customer churn* constituye un problema especialmente interesante porque reúne varios de los desafíos clásicos del aprendizaje supervisado. Entre ellos destacan la coexistencia de variables numéricas y categóricas, la necesidad de realizar procesos de limpieza y transformación de los datos, la presencia de clases desbalanceadas y la comparación entre modelos con distintos niveles de complejidad e interpretabilidad. Asimismo, diversos estudios destacan que la evaluación de los modelos debe considerar múltiples métricas de desempeño, tales como *Accuracy*, *Precision*, *Recall*, *F1-score* y *ROC-AUC*, ya que una única medida no resulta suficiente para caracterizar completamente el comportamiento de un clasificador, especialmente cuando el costo de no detectar un cliente con riesgo de abandono puede ser significativamente mayor que el de clasificar erróneamente a un cliente que permanecerá en la compañía **(Alotaibi, 2024)**.

Con el propósito de estudiar este problema, el presente trabajo utiliza dos conjuntos de datos públicos disponibles en Kaggle: **Customer Churn Dataset**, compuesto por aproximadamente 440.000 registros de entrenamiento y 64.000 registros de prueba, e **IBM Telco Customer Churn Dataset**, conformado por 7.043 clientes y 21 variables relacionadas con características demográficas, servicios contratados y antecedentes de facturación. La utilización de ambos conjuntos de datos permite evaluar la consistencia del desempeño de diferentes algoritmos de clasificación supervisada en escenarios con distintos tamaños muestrales y características de los datos, proporcionando un análisis comparativo más robusto sobre su capacidad de generalización y su aplicabilidad en problemas reales de predicción del abandono de clientes.

En este contexto, el presente estudio utiliza la predicción de *customer churn* como caso de estudio para analizar y comparar distintas técnicas de aprendizaje supervisado aplicadas a un problema real de negocio. Además de evaluar el desempeño predictivo de diferentes algoritmos de clasificación, se analizará cómo distintas estrategias para el tratamiento del desbalance de clases influyen en la capacidad de identificar clientes con riesgo de abandono. De esta forma, el estudio busca aportar evidencia que permita seleccionar modelos que no sólo presenten un buen desempeño estadístico, sino que también constituyan una herramienta útil para apoyar la toma de decisiones en procesos de retención de clientes.

## 4.1 Principios metodológicos

Con el propósito de asegurar la objetividad, reproducibilidad y validez del estudio, el desarrollo del análisis se basa en los siguientes principios metodológicos:

1. **Las decisiones metodológicas se sustentan en evidencia cuantitativa.** La selección del conjunto de datos, los algoritmos, las estrategias de balanceo y los hiperparámetros se realiza utilizando procedimientos de evaluación objetivos, evitando decisiones basadas únicamente en criterios subjetivos.

2. **La comparación entre modelos considera múltiples métricas de desempeño.** Dado que la predicción de *customer churn* corresponde a un problema de clasificación desbalanceada, se adopta el **PR-AUC** (*average precision*) como métrica de selección, por ser independiente del umbral de decisión y concentrarse en la clase minoritaria. La evaluación se complementa con once métricas adicionales —entre ellas *Precision*, *Recall*, *F1*, *F2*, *ROC-AUC*, *balanced accuracy*, especificidad, coeficiente de correlación de Matthews y *kappa* de Cohen— con el fin de caracterizar de forma integral el comportamiento de cada modelo.

3. **El tratamiento del desbalance de clases forma parte del proceso de evaluación.** Se analizan distintas estrategias de balanceo con el objetivo de determinar su efecto sobre la capacidad de identificar correctamente a los clientes con riesgo de abandono.

4. **La selección del modelo final considera tanto criterios estadísticos como prácticos.** Además del desempeño predictivo, se evalúan aspectos relacionados con la capacidad de generalización, la interpretabilidad y la utilidad del modelo como apoyo para la toma de decisiones en estrategias de retención de clientes. Cuando dos modelos resultan estadísticamente equivalentes, se aplica el principio de parsimonia.

5. **El conjunto de prueba se reserva para la evaluación final.** La totalidad de las decisiones metodológicas —selección de atributos, hiperparámetros, escalamiento, balanceo y elección del modelo— se adopta mediante validación cruzada sobre el conjunto de entrenamiento. El conjunto de prueba se emplea una única vez, con el propósito de reportar el desempeño del modelo seleccionado.

6. **Los análisis son reproducibles.** Se fija una semilla aleatoria común (`RANDOM_STATE = 42`) en todos los procedimientos que involucran aleatoriedad, incluidas las particiones, la búsqueda de hiperparámetros y los métodos de remuestreo.

# Fase 1 — Selección de datos

**Objetivo de la fase:** definir el problema y decidir, con evidencia, **sobre qué datos
se construye el modelo**.

## 1.1 Comprensión del problema

El *churn* (fuga de clientes) es un problema de **clasificación binaria desbalanceada**
con una asimetría de costos clara: **no detectar** a un cliente que se fuga (falso
negativo) cuesta mucho más que contactar innecesariamente a uno que se queda (falso
positivo). Esto define dos decisiones metodológicas que se arrastran a todo el proyecto:

- La métrica principal **no puede ser *accuracy*** (un modelo que predice "nadie se fuga"
  acertaría ~73% y sería inútil).
- Se usará **PR-AUC** (*average precision*) como métrica de selección, complementada con
  **Recall** y un **umbral operativo** calibrado al costo del negocio (§5.2).

## 1.2 Carga de los datos candidatos

```python
import kagglehub, os

kaggle_path = kagglehub.dataset_download("muhammadshahidazeem/customer-churn-dataset")
telco_path  = kagglehub.dataset_download("blastchar/telco-customer-churn")

df_train_raw = pd.read_csv(os.path.join(kaggle_path, "customer_churn_dataset-training-master.csv"))
df_test_raw  = pd.read_csv(os.path.join(kaggle_path, "customer_churn_dataset-testing-master.csv"))
df_telco_raw = pd.read_csv(os.path.join(telco_path,  "WA_Fn-UseC_-Telco-Customer-Churn.csv"))

print("Candidato A — 'Customer Churn':", df_train_raw.shape, "train |", df_test_raw.shape, "test")
print("Candidato B — 'Telco Customer Churn':", df_telco_raw.shape)
```

## 1.3 Auditoría de aptitud de los datos (decisión de selección)

Antes de modelar, KDD exige verificar que los datos **sirven** para el objetivo. Aplicamos
tres pruebas al candidato A, porque los resultados preliminares mostraron una señal
anómala: **AUC de validación = 1.0000** y F1 ≈ 0.999. En aprendizaje supervisado, un
desempeño perfecto casi nunca significa un buen modelo: significa **fuga de información
(*leakage*) o datos sintéticos generados por una regla**.

**Prueba 1 — ¿Es separable por una regla trivial?**
Si un árbol de profundidad 3 alcanza AUC ≈ 1.0, la etiqueta fue generada por una regla
determinista y el problema no tiene incertidumbre real que aprender.


```python
def split_xy(frame):
    frame = frame.dropna(subset=["Churn"]).copy()
    frame["Churn"] = frame["Churn"].astype(int)
    return frame.drop(columns=["CustomerID","Churn"]), frame["Churn"]

X_A, y_A = split_xy(df_train_raw)
X_A_test, y_A_test = split_xy(df_test_raw)

num_A = X_A.select_dtypes("number").columns.tolist()
tree = DecisionTreeClassifier(max_depth=3, random_state=RANDOM_STATE).fit(X_A[num_A], y_A)
auc_tree = roc_auc_score(y_A, tree.predict_proba(X_A[num_A])[:, 1])

print(f"AUC de un árbol de profundidad 3 (solo numéricas): {auc_tree:.4f}")
print("-> Si es ~1.0, la etiqueta se explica por una regla determinista.\n")
print("Regla aprendida:")
print(export_text(tree, feature_names=num_A, max_depth=2))
```

**Prueba 2 — ¿Coinciden las distribuciones de train y test?**
Validación adversarial: entrenamos un clasificador para distinguir el archivo de origen.
AUC ≫ 0.5 implica que **no** son intercambiables y que el test provisto no mide
generalización sobre la misma población.

```python
adv = pd.concat([X_A, X_A_test], ignore_index=True)
adv_y = np.r_[np.zeros(len(X_A)), np.ones(len(X_A_test))]
idx = np.random.RandomState(RANDOM_STATE).choice(len(adv), size=min(40000, len(adv)), replace=False)
Xadv, yadv = adv.iloc[idx], adv_y[idx]

na = Xadv.select_dtypes("number").columns.tolist()
ca = Xadv.select_dtypes(exclude="number").columns.tolist()
prep_adv = ColumnTransformer([
    ("n", Pipeline([("i",SimpleImputer(strategy="median")),("s",StandardScaler())]), na),
    ("c", Pipeline([("i",SimpleImputer(strategy="most_frequent")),
                    ("o",OneHotEncoder(handle_unknown="ignore"))]), ca)])
adv_clf = Pipeline([("p",prep_adv),
    ("m",RandomForestClassifier(n_estimators=200, n_jobs=-1, random_state=RANDOM_STATE))])
adv_auc = cross_val_score(adv_clf, Xadv, yadv, cv=3, scoring="roc_auc", n_jobs=-1).mean()

print(f"AUC adversarial (distinguir train de test): {adv_auc:.4f}")
print("   0.5 = misma distribución | ~1.0 = distribuciones distintas\n")
print("Tasa de fuga  train:", round(y_A.mean(), 4), "| test:", round(y_A_test.mean(), 4))
```

**Prueba 3 — ¿Qué predice el modelo en el test provisto?**
Si el *recall* es ≈ 1.0 con *precision* ≈ tasa base, el modelo está prediciendo
"fuga" para casi todos: la regla aprendida en train no transfiere al test.

```python
chk = Pipeline([("p", ColumnTransformer([
        ("n", Pipeline([("i",SimpleImputer(strategy="median")),("s",StandardScaler())]), num_A)])),
    ("m", LogisticRegression(max_iter=1000, random_state=RANDOM_STATE))]).fit(X_A[num_A], y_A)
pred_test = chk.predict(X_A_test[num_A])

print(f"Predice 'fuga' en el {pred_test.mean()*100:.1f}% del test provisto "
      f"(tasa real de fuga en test: {y_A_test.mean()*100:.1f}%)")
print(f"Recall test    = {recall_score(y_A_test, pred_test):.4f}")
print(f"Precision test = {precision_score(y_A_test, pred_test):.4f}")
```

### Decisión de la Fase 1 (argumentada)

| Evidencia | Resultado | Implicación |
|---|---|---|
| Árbol de profundidad 3 | AUC ≈ 1.0 | La etiqueta es **determinista**: dataset sintético generado por reglas |
| Validación adversarial | AUC ≈ 0.74 ≫ 0.5 | Train y test **no** comparten distribución |
| Predicciones en test | Recall ≈ 1.0, Precision ≈ tasa base | El modelo **colapsa** a predecir "fuga" |

**Resultado.** El candidato A **no es apto** como dataset de modelamiento: no existe
incertidumbre que aprender (cualquier modelo llega a ~1.0), y su test está corrupto por
*shift*. Los "resultados excelentes" (F1 = 0.999) son un **artefacto**, no un logro —
reportarlos como éxito sería un error metodológico grave.

> **Decisión:** el proyecto se desarrolla sobre el **candidato B (Telco Customer Churn)**,
> que presenta ruido realista, desbalance genuino (~26.5%) y features de negocio
> interpretables. El candidato A se conserva en el **Anexo (§5.5)** como *caso de estudio
> de auditoría de calidad de datos* — que es, en sí mismo, un resultado valioso del
> proceso KDD: **determinar que un conjunto de datos no es apto constituye un hallazgo del proceso.**

## 3.1 Ingeniería de atributos derivada del EDA

Cada atributo nuevo responde a un hallazgo concreto de la Fase 2 — no son arbitrarios:

| Atributo nuevo | Hallazgo que lo origina | Hipótesis de negocio |
|---|---|---|
| `MonthToMonth` | Hallazgo 4: `Contract` domina | Aísla el segmento de máximo riesgo |
| `tenure_group`, `NewCustomer` | Hallazgo 3: fuga temprana | Captura el efecto no lineal de la antigüedad |
| `AvgChargesPerMonth` | Hallazgo 5: multicolinealidad | Normaliza gasto sin depender de `tenure` |
| `Charges_gap` | Hallazgo 5 | Desvío entre lo facturado y lo esperado |
| `TotalServices` | Hallazgo 4: add-ons | *Engagement*: más servicios ⇒ más costo de cambio |
| `FiberNoSecurity` | Hallazgo 4 (Fiber + sin seguridad) | Interacción del segmento con mayor tasa de abandono |
| `AutoPay` | Hallazgo 4: `PaymentMethod` | Pago automático ⇒ fricción de salida |
| `Streaming` | Hallazgo 4 | Uso de servicios de entretención |

```python
SERVICES = ["OnlineSecurity","OnlineBackup","DeviceProtection",
            "TechSupport","StreamingTV","StreamingMovies"]

def fe_telco(X):
    """Atributos derivados. Cada uno justificado por un hallazgo del EDA (§2)."""
    X = X.copy()
    t = X["tenure"].replace(0, np.nan)
    X["TotalServices"]      = (X[SERVICES] == "Yes").sum(axis=1)       # Hallazgo 4
    X["AvgChargesPerMonth"] = (X["TotalCharges"] / t).fillna(X["MonthlyCharges"])  # Hallazgo 5
    X["Charges_gap"]        = X["MonthlyCharges"]*X["tenure"] - X["TotalCharges"]  # Hallazgo 5
    X["NewCustomer"]        = (X["tenure"] <= 6).astype(int)           # Hallazgo 3
    X["tenure_group"]       = pd.cut(X["tenure"], [-1,12,24,48,100],
                                     labels=["0-1a","1-2a","2-4a","4a+"]).astype(str)
    X["AutoPay"]            = X["PaymentMethod"].str.contains("automatic").astype(int)
    X["FiberNoSecurity"]    = ((X["InternetService"]=="Fiber optic") &
                               (X["OnlineSecurity"]=="No")).astype(int) # Hallazgo 4
    X["MonthToMonth"]       = (X["Contract"]=="Month-to-month").astype(int)
    X["Streaming"]          = ((X["StreamingTV"]=="Yes") |
                               (X["StreamingMovies"]=="Yes")).astype(int)
    return X.replace([np.inf, -np.inf], np.nan)

X_raw = telco.drop(columns=["Churn"])
X_all = fe_telco(X_raw)
y_all = telco["Churn"]

NUM = X_all.select_dtypes("number").columns.tolist()
CAT = X_all.select_dtypes(exclude="number").columns.tolist()
print(f"Atributos: {X_raw.shape[1]} originales -> {X_all.shape[1]} tras FE")
print(f"   numéricas: {len(NUM)} | categóricas: {len(CAT)}")
print("Nuevos:", [c for c in X_all.columns if c not in X_raw.columns])
```

## 3.2 Análisis de relevancia de variables

> Esta sección responde: **¿qué variables importan realmente para la etiqueta?**

Ningún método por sí solo es confiable: los univariados ignoran interacciones, la
importancia por *impurity* se sesga hacia variables de alta cardinalidad, y los
coeficientes se distorsionan con multicolinealidad (Hallazgo 5). Por eso usamos
**cinco métodos complementarios** y los consolidamos en un **ranking por consenso**
(promedio de rangos normalizados, tipo *Borda count*):

| Método | Tipo | Qué captura | Limitación |
|---|---|---|---|
| **Información mutua (MI)** | Univariado | Dependencia no lineal con la etiqueta | Ignora interacciones |
| **ANOVA F / χ²** | Univariado | Significancia estadística (p-valor) | Solo relación marginal |
| **corr. biserial / Cramér's V** | Univariado | Fuerza y dirección de asociación | Solo relación monótona |
| **Importancia Random Forest** | Multivariado | Contribución en presencia de otras | Sesgo por cardinalidad |
| **Permutation importance** | Multivariado | Caída real de PR-AUC al destruir la variable | Costosa; diluye con correlación |

```python
def cramers_v(x, y):
    ct = pd.crosstab(x, y); c2 = chi2_contingency(ct)[0]
    n = ct.values.sum(); r, k = ct.shape
    return np.sqrt((c2/n) / max(min(r-1, k-1), 1))

def relevancia_univariada(X, y, num, cat, random_state=RANDOM_STATE):
    Xi = X.copy()
    for c in num: Xi[c] = Xi[c].fillna(Xi[c].median())
    for c in cat: Xi[c] = Xi[c].fillna(Xi[c].mode()[0]).astype(str)
    rows = []
    mi_n = mutual_info_classif(Xi[num], y, random_state=random_state)
    F, pF = f_classif(Xi[num], y)
    for i, c in enumerate(num):
        rho, _ = pointbiserialr(y, Xi[c])
        rows.append({"variable":c, "tipo":"numérica", "MI":mi_n[i], "asociación":abs(rho),
                     "estadístico":F[i], "p_valor":pF[i], "test":"ANOVA F"})
    enc = OrdinalEncoder().fit_transform(Xi[cat])
    mi_c = mutual_info_classif(enc, y, discrete_features=True, random_state=random_state)
    for i, c in enumerate(cat):
        c2, p, _, _ = chi2_contingency(pd.crosstab(Xi[c], y))
        rows.append({"variable":c, "tipo":"categórica", "MI":mi_c[i],
                     "asociación":cramers_v(Xi[c], y), "estadístico":c2,
                     "p_valor":p, "test":"Chi²"})
    return pd.DataFrame(rows).set_index("variable")

rel = relevancia_univariada(X_all, y_all, NUM, CAT)
print("Variables SIN asociación significativa con la etiqueta (p > 0.05):")
ns = rel[rel["p_valor"] > 0.05]
display(ns[["tipo","MI","asociación","p_valor"]].round(4) if len(ns)
        else "  (ninguna: todas muestran asociación significativa)")
print("\nTop 12 por Información Mutua:")
display(rel.sort_values("MI", ascending=False).head(12).round(4))
```

### 3.2.1 Métodos multivariados y ranking consolidado

```python
def build_prep(num, cat, scaler=StandardScaler()):
    """Preprocesador. El escalador queda expuesto como hiperparámetro (ver §3.4)."""
    return ColumnTransformer([
        ("num", Pipeline([("i", SimpleImputer(strategy="median")), ("s", scaler)]), num),
        ("cat", Pipeline([("i", SimpleImputer(strategy="most_frequent")),
                          ("o", OneHotEncoder(handle_unknown="ignore"))]), cat)])

def agrupar_ohe(imps, feat_names, originales):
    """Devuelve importancias de columnas one-hot a su variable original (interpretable)."""
    s = pd.Series(0.0, index=originales)
    for fn, v in zip(feat_names, imps):
        base = fn.split("__", 1)[1]
        cand = [o for o in originales if base == o or base.startswith(o + "_")]
        if cand: s[max(cand, key=len)] += v
    return s

# split estratificado (se reutiliza en toda la Fase 4)
X_tr, X_te, y_tr, y_te = train_test_split(
    X_all, y_all, test_size=0.25, stratify=y_all, random_state=RANDOM_STATE)
print(f"Train: {X_tr.shape} | Test: {X_te.shape}")

prep0 = build_prep(NUM, CAT)
# (a) importancia Random Forest
rf = Pipeline([("prep", prep0), ("m", RandomForestClassifier(
    n_estimators=400, class_weight="balanced", n_jobs=-1,
    random_state=RANDOM_STATE))]).fit(X_tr, y_tr)
fnames = rf.named_steps["prep"].get_feature_names_out()
rf_imp = agrupar_ohe(rf.named_steps["m"].feature_importances_, fnames, list(X_all.columns))

# (b) permutation importance (sobre test, mide caída real de PR-AUC)
perm = permutation_importance(rf, X_te, y_te, n_repeats=10,
    scoring="average_precision", random_state=RANDOM_STATE, n_jobs=-1)
perm_imp = pd.Series(perm.importances_mean, index=X_all.columns)

# --- consolidación por consenso de rangos ---
tabla = rel.copy()
tabla["RF_imp"] = rf_imp
tabla["Perm_imp"] = perm_imp
METODOS = ["MI", "asociación", "RF_imp", "Perm_imp"]
rangos = tabla[METODOS].rank(ascending=False, pct=True)
tabla["score_consenso"] = rangos.mean(axis=1)
tabla = tabla.sort_values("score_consenso")
tabla.insert(0, "ranking", range(1, len(tabla)+1))

print("\n=== RANKING CONSOLIDADO DE RELEVANCIA (menor score = más relevante) ===")
display(tabla[["ranking","tipo","MI","asociación","RF_imp","Perm_imp","score_consenso"]].round(4))
```

```python
# visualización: acuerdo entre métodos
fig, ax = plt.subplots(1, 2, figsize=(16, 7))
top = tabla.head(15).index[::-1]
hm = rangos.loc[top][METODOS]
sns.heatmap(hm, annot=True, fmt=".2f", cmap="RdYlGn_r", ax=ax[0],
            cbar_kws={"label":"rango normalizado (0=más relevante)"})
ax[0].set_title("Acuerdo entre métodos — Top 15 variables")
tabla.loc[top, "Perm_imp"].plot(kind="barh", ax=ax[1], color=sns.color_palette("deep")[3])
ax[1].axvline(0, color="black", lw=.8)
ax[1].set_title("Permutation importance (caída de PR-AUC)")
ax[1].set_xlabel("importancia")
plt.tight_layout(); plt.show()

irrelevantes = tabla[(tabla["Perm_imp"] <= 0) & (tabla["score_consenso"] > 0.5)].index.tolist()
print(f"Candidatas a descarte (aporte nulo/negativo y bajo consenso): {irrelevantes}")
```

> **Lectura del ranking.** El **consenso entre métodos independientes** es el argumento:
> cuando MI (univariado), Random Forest (multivariado) y *permutation importance*
> (impacto causal en PR-AUC) coinciden en las mismas variables al tope, el resultado es robusto y no un artefacto de un método particular.
>
> El **heatmap** muestra dónde los métodos **discrepan**, que es igual de informativo: una
> variable bien rankeada en MI pero mal en *permutation* suele ser **redundante** — aporta
> información que otra variable correlacionada ya entrega (típicamente el caso de
> `TotalCharges` vs `tenure`, Hallazgo 5).

## 3.3 Selección del conjunto de variables (decisión)

El ranking **ordena**, pero no dice **cuántas** variables conservar. Para eso usamos
**RFECV** (eliminación recursiva con validación cruzada), que determina el número óptimo
maximizando PR-AUC. Luego **contrastamos empíricamente** tres conjuntos candidatos.

```python
Xtr_enc = prep0.fit_transform(X_tr, y_tr)
Xtr_enc = Xtr_enc.toarray() if hasattr(Xtr_enc, "toarray") else Xtr_enc

rfecv = RFECV(
    estimator=RandomForestClassifier(n_estimators=150, n_jobs=-1, random_state=RANDOM_STATE),
    step=3, cv=StratifiedKFold(3, shuffle=True, random_state=RANDOM_STATE),
    scoring="average_precision", min_features_to_select=5, n_jobs=-1).fit(Xtr_enc, y_tr)

plt.figure(figsize=(9, 4.5))
n_feat = range(1, len(rfecv.cv_results_["mean_test_score"])+1)
plt.errorbar([rfecv.min_features_to_select + i*rfecv.step for i in range(len(n_feat))],
             rfecv.cv_results_["mean_test_score"], yerr=rfecv.cv_results_["std_test_score"],
             marker="o", ms=3)
plt.axvline(rfecv.n_features_, color="red", ls="--", label=f"óptimo = {rfecv.n_features_}")
plt.xlabel("nº de atributos (codificados)"); plt.ylabel("PR-AUC (CV)")
plt.title("RFECV — ¿cuántos atributos conservar?"); plt.legend(); plt.show()
print(f"RFECV -> {rfecv.n_features_} atributos óptimos de {Xtr_enc.shape[1]} codificados")
```

```python
# --- contraste empírico de conjuntos candidatos ---
TOP_K = 15
set_todas = list(X_all.columns)
set_top   = tabla.head(TOP_K).index.tolist()
set_sinirr = [c for c in set_todas if c not in irrelevantes]

cv_sel = StratifiedKFold(CV_FOLDS, shuffle=True, random_state=RANDOM_STATE)
def eval_set(cols, nombre):
    n = [c for c in cols if c in NUM]; c_ = [c for c in cols if c in CAT]
    pipe = Pipeline([("prep", build_prep(n, c_)),
        ("m", LogisticRegression(max_iter=2000, class_weight="balanced",
                                 random_state=RANDOM_STATE))])
    s = cross_val_score(pipe, X_all[cols], y_all, cv=cv_sel,
                        scoring="average_precision", n_jobs=-1)
    return {"conjunto":nombre, "n_variables":len(cols),
            "PR-AUC":s.mean(), "sd":s.std()}

comp = pd.DataFrame([
    eval_set(set_todas,  "A) Todas (post-FE)"),
    eval_set(set_sinirr, "B) Sin irrelevantes"),
    eval_set(set_top,    f"C) Top-{TOP_K} del consenso"),
]).set_index("conjunto").sort_values("PR-AUC", ascending=False)
display(comp.round(4))

FEATURES = comp.index[0]
FEATURES = {"A) Todas (post-FE)":set_todas, "B) Sin irrelevantes":set_sinirr,
            f"C) Top-{TOP_K} del consenso":set_top}[FEATURES]
NUM_F = [c for c in FEATURES if c in NUM]; CAT_F = [c for c in FEATURES if c in CAT]
print(f"\nDECISIÓN §3.3 -> conjunto '{comp.index[0]}' con {len(FEATURES)} variables")
print("   (elegido por mayor PR-AUC de CV; ante empate <0.005 preferir el más pequeño)")
```

> **Decisión de la Fase 3.** El conjunto se elige **por evidencia empírica** (PR-AUC de
> CV), no por intuición. Que "todas las variables" gane suele indicar que los modelos
> regularizados ya manejan la redundancia por sí solos; que gane un subconjunto indica que
> el ruido estaba perjudicando. **Ambos resultados son válidos y quedan documentados.**

## 3.4 Escalamiento: justificación de su determinación empírica

Comparar escaladores con un modelo fijo (como se hacía antes) es metodológicamente débil:
el escalador óptimo **depende del modelo**. La forma correcta es tratarlo como **un
hiperparámetro más** y dejar que la búsqueda de la Fase 4 lo resuelva por modelo.

**Un argumento importante que ahorra cómputo:** los modelos basados en **árboles**
(RF, ExtraTrees, GB, HistGB, XGBoost, LightGBM) dividen por umbrales del tipo
`x ≤ c`. Cualquier transformación **monótona** (Standard, Robust, MinMax, Yeo-Johnson)
preserva el orden de los datos, por lo que **los cortes no cambian**: escalar es
**matemáticamente irrelevante** para ellos. En cambio, los modelos basados en
**distancia o gradiente** (kNN, SVM, MLP, LogReg) sí son sensibles.

Por eso la grilla asigna escaladores **solo a los modelos que los necesitan** — decisión
justificada, no arbitraria. Lo verificamos empíricamente:

# Fase 4 — Minería de datos (modelamiento)

## 4.1 Protocolo experimental

### 4.1.1 Presupuesto adaptativo de búsqueda (corrección metodológica)

En la iteración anterior el presupuesto era **fijo** (`n_iter=12`) para todos los modelos.
Eso fue un **error de diseño**, y es la causa de que la calidad bajara respecto de la
iteración previa (donde el mejor modelo recibía `n_iter=30` sobre una grilla enfocada).
**No fue por falta de balanceo** — el remuestreo ya estaba en la grilla; lo que faltó fue
**exploración**.

**El argumento correcto no es la "cobertura %".** En búsqueda aleatoria no se necesita
recorrer el espacio: basta con **sortear suficientes veces** para caer en la región buena.
Bergstra & Bengio (2012) lo formalizan — si el **q%** superior de las configuraciones es
"bueno", la probabilidad de encontrar al menos una en `n` sorteos es:

$$P(\text{éxito}) = 1-(1-q)^{n}$$

Para el 5% superior (q = 0.05):

| `n_iter` | P(encontrar una config. del top-5%) |
|---|---|
| 12 (iteración anterior) | **46%** ← equivalente al azar |
| 20 | 64% |
| 30 (`FAST_MODE=True`) | **79%** |
| 70 (`FAST_MODE=False`) | **97%** |

Con `n_iter=12` había **más de un 50% de probabilidad de no encontrar nunca una buena
configuración**. Eso explica exactamente la pérdida de calidad observada.

**Corrección aplicada:**

- Grillas pequeñas (≤ `EXHAUSTIVE_MAX`) → **`GridSearchCV` exhaustivo** (garantía total).
- Grillas grandes → `RandomizedSearchCV` con `n_iter ∝ √(espacio)`, acotado entre
  `MIN_ITER` y `MAX_ITER`, y con tope adicional para los modelos computacionalmente caros.
- Se reporta **P(top-5%)** de cada modelo, que es la garantía estadística real de la búsqueda.


```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import (make_scorer, matthews_corrcoef, fbeta_score,
                             cohen_kappa_score, balanced_accuracy_score)

# --- Presupuesto ---
EXHAUSTIVE_MAX = 60          # grillas <= a esto se recorren completas
BUDGET_K       = 1.6         # n_iter ~ BUDGET_K * sqrt(espacio)
MIN_ITER, MAX_ITER = 20, (30 if FAST_MODE else 70)
# modelos caros: se les acota el presupuesto (no el espacio)
TOPE_POR_COSTO = {"SVM (RBF)":15, "MLP":15, "GradientBoosting":20, "ExtraTrees":25,
                  "Random Forest":25}

def tamaño_grilla(grid):
    """Soporta dict o LISTA de dicts (grillas condicionales)."""
    if isinstance(grid, list):
        return sum(int(np.prod([len(v) for v in g.values()])) for g in grid)
    return int(np.prod([len(v) for v in grid.values()]))

def p_top(n, q=0.05):
    """P(hallar >=1 config del top q%) en n sorteos aleatorios (Bergstra & Bengio)."""
    return 1 - (1-q)**n

def presupuesto(nombre, grid):
    """Devuelve (estrategia, n_iter, garantía) para el modelo."""
    esp = tamaño_grilla(grid)
    if esp <= EXHAUSTIVE_MAX:
        return "grid", esp, 1.0          # exhaustivo: garantía total
    n = int(np.clip(BUDGET_K*np.sqrt(esp), MIN_ITER, MAX_ITER))
    n = min(n, TOPE_POR_COSTO.get(nombre, MAX_ITER))
    return "random", n, p_top(n)

# --- 11 métricas: un solo número esconde el comportamiento real ---
SCORING = {
    "PR-AUC":        "average_precision",
    "ROC-AUC":       "roc_auc",
    "F1":            "f1",
    "F2":            make_scorer(fbeta_score, beta=2),   # pondera Recall 4x sobre Precision
    "Precision":     "precision",
    "Recall":        "recall",
    "Accuracy":      "accuracy",
    "BalancedAcc":   "balanced_accuracy",
    "Especificidad": make_scorer(recall_score, pos_label=0),
    "MCC":           make_scorer(matthews_corrcoef),
    "Kappa":         make_scorer(cohen_kappa_score),
}
print("Métricas que se registrarán en CADA configuración probada:")
for k in SCORING: print("   -", k)
print(f"\nPresupuesto: exhaustivo si espacio<={EXHAUSTIVE_MAX}; "
      f"si no, n_iter=clip({BUDGET_K}*sqrt(espacio), {MIN_ITER}, {MAX_ITER})")
print("\nGarantía estadística de la búsqueda aleatoria — P(hallar config. del top-5%):")
for n in [12, 20, 30, 50, 70]:
    marca = "  <- iteración anterior (insuficiente)" if n==12 else ""
    print(f"   n_iter={n:3d} -> {p_top(n)*100:5.1f}%{marca}")
```

> **Por qué 11 métricas y no solo F1.** En un problema desbalanceado cada métrica cuenta
> una parte distinta de la historia:
>
> - **PR-AUC / ROC-AUC**: calidad del *ranking* de probabilidades, sin depender del umbral.
> - **Recall**: % de fugados que efectivamente detectamos → **la métrica del negocio**.
> - **Precision**: % de aciertos entre los contactados → costo de la campaña.
> - **F2**: como F1 pero pondera *recall* 4× más que *precision*, coherente con que un
>   falso negativo cuesta 5× un falso positivo (§1.1).
> - **Accuracy**: se incluye para **mostrar que engaña** (el trivial logra 73.5%).
> - **BalancedAcc / MCC / Kappa**: robustas al desbalance; MCC solo es alto si las
>   **cuatro** celdas de la matriz de confusión son buenas.
> - **Especificidad**: % de clientes fieles correctamente no molestados.

## 4.2 Balanceo de clases (muestreo)

> Esta sección hace **explícito y visible** el muestreo. En la versión anterior el
> remuestreo sí se aplicaba, pero estaba *dentro* del diccionario de la grilla y no se
> mostraba por pantalla, lo que impedía auditarlo.

### 4.2.1 Efecto de cada técnica sobre la composición de clases

Aplicamos cada técnica al conjunto de entrenamiento y mostramos **cuántos casos de cada
clase quedan**. Las familias son distintas conceptualmente:

| Familia | Técnicas | Qué hace |
|---|---|---|
| **Sobremuestreo** | `RandomOverSampler`, `SMOTE`, `BorderlineSMOTE`, `ADASYN` | Crea casos de la clase minoritaria (replicando o interpolando vecinos) |
| **Submuestreo** | `RandomUnderSampler`, `TomekLinks` | Elimina casos de la clase mayoritaria |
| **Híbridas** | `SMOTEENN`, `SMOTETomek` | Sobremuestrean y luego limpian el ruido de la frontera |
| **Ponderación** | `class_weight`, `scale_pos_weight` | No toca los datos: penaliza más el error en la minoritaria |

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import TomekLinks

prep_fit = build_prep(NUM_F, CAT_F).fit(X_tr[FEATURES], y_tr)
Xenc_tr = prep_fit.transform(X_tr[FEATURES])
Xenc_tr = Xenc_tr.toarray() if hasattr(Xenc_tr, "toarray") else Xenc_tr

TECNICAS = {
    "Original (sin balancear)": None,
    "RandomOverSampler":  RandomOverSampler(random_state=RANDOM_STATE),
    "SMOTE":              SMOTE(random_state=RANDOM_STATE),
    "BorderlineSMOTE":    BorderlineSMOTE(random_state=RANDOM_STATE),
    "ADASYN":             ADASYN(random_state=RANDOM_STATE),
    "RandomUnderSampler": RandomUnderSampler(random_state=RANDOM_STATE),
    "TomekLinks":         TomekLinks(),
    "SMOTEENN":           SMOTEENN(random_state=RANDOM_STATE),
    "SMOTETomek":         SMOTETomek(random_state=RANDOM_STATE),
}
filas = []
for nombre, tec in TECNICAS.items():
    yy = y_tr.values if tec is None else tec.fit_resample(Xenc_tr, y_tr)[1]
    c = Counter(yy); tot = sum(c.values())
    filas.append({"Técnica":nombre, "No fuga (0)":c[0], "Fuga (1)":c[1],
                  "Total":tot, "Ratio 0:1":round(c[0]/c[1],2),
                  "% minoritaria":round(100*c[1]/tot,1),
                  "Δ vs original":tot-len(y_tr)})
balance = pd.DataFrame(filas).set_index("Técnica")
print("=== COMPOSICIÓN DE CLASES DESPUÉS DE CADA TÉCNICA (sobre el TRAIN) ===")
display(balance)

fig, ax = plt.subplots(1, 2, figsize=(16, 5))
balance[["No fuga (0)","Fuga (1)"]].plot(kind="barh", stacked=True, ax=ax[0],
    color=[sns.color_palette("deep")[0], sns.color_palette("deep")[3]])
ax[0].set_title("Nº de casos por clase según técnica"); ax[0].set_xlabel("casos")
balance["% minoritaria"].plot(kind="barh", ax=ax[1], color=sns.color_palette("deep")[2])
ax[1].axvline(50, color="green", ls="--", label="balance perfecto (50%)")
ax[1].axvline(balance.loc["Original (sin balancear)","% minoritaria"], color="red",
              ls=":", label="original")
ax[1].set_title("% de la clase minoritaria"); ax[1].legend()
plt.tight_layout(); plt.show()
```

> **Lectura.** `SMOTE`/`RandomOver`/`SMOTETomek` igualan ambas clases al 50% **agregando**
> casos sintéticos (el total crece). `RandomUnderSampler` también iguala pero **descartando**
> datos (el total se desploma — pierde información). `TomekLinks` apenas cambia el balance
> porque solo limpia pares frontera. `SMOTEENN` es el más agresivo: sobremuestrea y luego
> **borra** tanto ruido que puede **invertir** el desbalance.
>
> **Ninguna es universalmente mejor** — por eso la decisión no se toma aquí, sino que se
> resuelve empíricamente en §4.2.3 y se optimiza por modelo en §4.4.

### 4.2.2 Consideración metodológica: punto de aplicación del muestreo

Este es el error más frecuente al usar SMOTE, y produce resultados espectaculares pero
**falsos**. Si se remuestrea **antes** de partir en *folds*, los casos sintéticos generados
a partir de un cliente terminan en *train* mientras ese cliente queda en *validación*: el
modelo evalúa sobre información que ya vio. Lo demostramos midiendo la inflación:

```python
cv_search = StratifiedKFold(CV_FOLDS, shuffle=True, random_state=RANDOM_STATE)
modelo_demo = LogisticRegression(max_iter=2000, random_state=RANDOM_STATE)

# (A) INCORRECTO: SMOTE aplicado a todo el train antes de la validación cruzada
Xs, ys = SMOTE(random_state=RANDOM_STATE).fit_resample(Xenc_tr, y_tr)
pr_mal = cross_val_score(clone(modelo_demo), Xs, ys, cv=cv_search,
                         scoring="average_precision", n_jobs=-1).mean()

# (B) CORRECTO: SMOTE dentro del pipeline -> se aplica solo al fold de entrenamiento
pipe_ok = ImbPipeline([("prep", build_prep(NUM_F, CAT_F)),
                       ("samp", SMOTE(random_state=RANDOM_STATE)),
                       ("m", clone(modelo_demo))])
pr_bien = cross_val_score(pipe_ok, X_tr[FEATURES], y_tr, cv=cv_search,
                          scoring="average_precision", n_jobs=-1).mean()

print("="*66)
print(f"(A) SMOTE ANTES del split  -> PR-AUC = {pr_mal:.4f}   <-- FALSO (fuga)")
print(f"(B) SMOTE DENTRO del fold  -> PR-AUC = {pr_bien:.4f}   <-- honesto")
print(f"    Inflación artificial   = {pr_mal-pr_bien:+.4f} ({(pr_mal/pr_bien-1)*100:.0f}% de mentira)")
print("="*66)
print("Por eso TODO el remuestreo de este notebook usa imblearn.Pipeline:")
print("el sampler solo actúa en fit() del fold de entrenamiento, nunca en el de validación.")
```

> **Resultado clave.** La diferencia es enorme. Un informe que reporte la variante (A)
> mostraría un modelo "excelente" que en producción fracasaría. Todo este notebook usa
> `imblearn.Pipeline`, que garantiza que el muestreo ocurra **dentro** de cada *fold*.
> Es la misma lección de la Fase 1: **un resultado excesivamente favorable debe interpretarse como una señal de alerta metodológica.**

### 4.2.3 Comparación de técnicas de balanceo con métricas completas

Ahora comparamos todas las técnicas **correctamente evaluadas** (dentro del *fold*),
con las 11 métricas, sobre un modelo fijo para aislar el efecto del balanceo.

> **Los tres hallazgos del balanceo:**
>
> 1. **El balanceo transforma *recall*, no el *ranking*.** Todas las estrategias mueven
>    fuertemente Recall y Precision, pero el **PR-AUC apenas cambia**. Razón: PR-AUC mide
>    la calidad del **ordenamiento** de probabilidades, y el remuestreo básicamente
>    **desplaza el umbral implícito** — no enseña al modelo a discriminar mejor.
> 2. **Existe un trade-off inevitable.** Subir el *recall* de la clase minoritaria baja la
>    *precision*: se contactan más clientes fieles por error. La pregunta correcta no es
>    "¿corresponde balancear?" sino "**qué incremento de *recall* justifica el costo de un falso positivo**" — cuestión abordada en el
>    umbral por costo (§5.2).
> 3. **`class_weight` suele igualar a SMOTE** a una fracción del costo, sin inventar datos
>    sintéticos. Por eso ambos compiten en la búsqueda (§4.4).
>
> **Resultado metodológico:** no se fija una técnica *a priori*. Todas entran a la
> búsqueda como hiperparámetro para que **cada modelo** encuentre la suya.

## 4.4 Ejecución de la búsqueda

Se recorre cada modelo con su presupuesto. Para **cada configuración probada** se registran
las **11 métricas**, de modo que el resultado no es un número suelto sino una tabla
auditable de experimentos.

> ⏱️ `FAST_MODE=True` (§0) ≈ 10–15 min. `FAST_MODE=False` amplía el presupuesto (≈ 30–45 min).

```python
def nombre_componente(obj):
    if isinstance(obj, str): return "Ninguno"
    n = type(obj).__name__.replace("Scaler","").replace("Transformer","")
    ss = getattr(obj, "sampling_strategy", None)
    if isinstance(ss, float): n += f"({ss})"
    return n

def fmt_hp(params):
    """Hiperparámetros del modelo, legibles."""
    return {k.replace("m__",""): (round(v,4) if isinstance(v,float) else v)
            for k, v in sorted(params.items()) if k.startswith("m__")}

resultados, mejores, experimentos = [], {}, []
t_ini = time.time()

for nombre, (est, grid) in POOL.items():
    t0 = time.time()
    estrategia, n_iter_m, garantia = presupuesto(nombre, grid)
    pipe = ImbPipeline([("prep", build_prep(NUM_F, CAT_F)),
                        ("samp", "passthrough"), ("m", est)])
    common = dict(scoring=SCORING, refit="PR-AUC", cv=cv_search,
                  n_jobs=-1, error_score=np.nan)
    if estrategia == "grid":
        bs = GridSearchCV(pipe, grid, **common)
    else:
        bs = RandomizedSearchCV(pipe, grid, n_iter=n_iter_m,
                                random_state=RANDOM_STATE, **common)
    bs.fit(X_tr[FEATURES], y_tr)

    cvr, bi = bs.cv_results_, bs.best_index_
    fila = {"Modelo": nombre}
    fila.update({k: cvr[f"mean_test_{k}"][bi] for k in SCORING})
    fila.update({
        "PR-AUC sd": cvr["std_test_PR-AUC"][bi],
        "escalador": nombre_componente(bs.best_params_.get("prep__num__s","passthrough")),
        "muestreo":  nombre_componente(bs.best_params_.get("samp","passthrough")),
        "hiperparámetros": fmt_hp(bs.best_params_),
        "P(top-5%)": round(garantia,3), "seg": round(time.time()-t0,1)})
    resultados.append(fila)
    mejores[nombre] = bs.best_estimator_

    # guardar TODAS las configuraciones probadas
    ex = pd.DataFrame({k: cvr[f"mean_test_{k}"] for k in SCORING})
    ex.insert(0, "Modelo", nombre)
    ex["escalador"] = [nombre_componente(p.get("prep__num__s","passthrough")) for p in cvr["params"]]
    ex["muestreo"]  = [nombre_componente(p.get("samp","passthrough")) for p in cvr["params"]]
    ex["hiperparámetros"] = [str(fmt_hp(p)) for p in cvr["params"]]
    experimentos.append(ex)

    print(f"{nombre:22s} PR-AUC={fila['PR-AUC']:.4f} F1={fila['F1']:.4f} "
          f"Rec={fila['Recall']:.4f} | {fila['escalador']:9s} | "
          f"{fila['muestreo']:18s} | {len(cvr['params']):3d} cfg ({time.time()-t0:5.1f}s)")

leaderboard = pd.DataFrame(resultados).set_index("Modelo").sort_values("PR-AUC", ascending=False)
EXPERIMENTOS = pd.concat(experimentos, ignore_index=True).sort_values("PR-AUC", ascending=False)
print(f"\nTiempo total: {(time.time()-t_ini)/60:.1f} min | "
      f"configuraciones evaluadas: {len(EXPERIMENTOS)}")
```

# Fase 5 — Interpretación y evaluación

## 5.1 Selección del modelo con mejor desempeño: contraste estadístico de las diferencias

El *leaderboard* de la §4.4 ordena por PR-AUC medio, pero **una diferencia de 0.004 entre
dos modelos puede ser puro azar del particionado**. Elegir por la media sin verificarlo es el error más común en comparación de modelos.

**Protocolo de decisión (en orden):**

1. **Re-evaluar a los finalistas** con `RepeatedStratifiedKFold` (5 folds × 3 repeticiones
   = **15 mediciones pareadas** por modelo), sobre exactamente las mismas particiones.
2. **Prueba de Wilcoxon pareada** de cada finalista contra el líder. Es la prueba
   recomendada por Demšar (2006) para comparar clasificadores: no asume normalidad y
   respeta el pareo por *fold*.
3. **Regla de decisión:** si un modelo **no es significativamente peor** que el líder
   (p ≥ 0.05), se considera **estadísticamente equivalente**. Entre los equivalentes se
   elige el más **simple** (navaja de Occam), porque a igual desempeño el modelo simple es
   más rápido, interpretable y estable en producción.

```python
N_FINALISTAS = 5
finalistas = leaderboard.index[:N_FINALISTAS].tolist()

# complejidad: 1=lineal simple ... 5=ensemble de ensembles
COMPLEJIDAD = {"Naive Bayes":1,"LDA":1,"QDA":2,"LogReg":1,"Árbol de Decisión":2,
    "kNN":2,"SVM (RBF)":3,"Random Forest":3,"ExtraTrees":3,"GradientBoosting":4,
    "HistGradientBoosting":4,"XGBoost":4,"LightGBM":4,"MLP":4,
    "Voting (soft)":5,"Stacking":5}

detalle_finalistas = []
rcv = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=RANDOM_STATE)
scores = {}
for n in finalistas:
    cvr = cross_validate(clone(mejores[n]), X_all[FEATURES], y_all, cv=rcv,
                         scoring=SCORING, n_jobs=-1)
    scores[n] = cvr["test_PR-AUC"]
    detalle_finalistas.append({"Modelo":n, **{k: cvr[f"test_{k}"].mean() for k in SCORING}})
    print(f"{n:22s} PR-AUC={scores[n].mean():.4f}+/-{scores[n].std():.4f} "
          f"F1={cvr['test_F1'].mean():.4f} Recall={cvr['test_Recall'].mean():.4f} "
          f"MCC={cvr['test_MCC'].mean():.4f}")

S = pd.DataFrame(scores)
líder = S.mean().idxmax()

print(f"\n{'='*70}\nPRUEBA DE WILCOXON PAREADA — líder: {líder}")
print(f"H0: no hay diferencia de PR-AUC vs. el líder\n{'='*70}")
equivalentes = [líder]
for c in S.columns:
    if c == líder: continue
    stat, p = wilcoxon(S[líder], S[c])
    dif = S[líder].mean() - S[c].mean()
    if p >= 0.05:
        equivalentes.append(c); veredicto = "EQUIVALENTE (no se rechaza H0)"
    else:
        veredicto = "significativamente peor"
    print(f"  vs {c:22s} Δ={dif:+.4f}  p={p:.4f}  -> {veredicto}")

print("\n=== FINALISTAS CON LAS 11 MÉTRICAS (CV repetida 5x3) ===")
display(pd.DataFrame(detalle_finalistas).set_index("Modelo").round(4))

cmp_tab = pd.DataFrame({"PR-AUC medio":S.mean(), "sd":S.std(),
    "complejidad":[COMPLEJIDAD.get(c,3) for c in S.columns],
    "equivalente al líder":[c in equivalentes for c in S.columns]})
display(cmp_tab.sort_values("PR-AUC medio", ascending=False).round(4))
```

```python
# --- decisión final: entre los equivalentes, el más simple ---
MEJOR_MODELO = min(equivalentes, key=lambda c: (COMPLEJIDAD.get(c, 3), -S[c].mean()))
modelo_final = mejores[MEJOR_MODELO]

print("="*70); print("ARGUMENTOS DE LA DECISIÓN".center(70)); print("="*70)
print(f"1. Líder por PR-AUC medio (15 mediciones): '{líder}' con {S[líder].mean():.4f}.")
print(f"2. Estadísticamente equivalentes al líder (Wilcoxon, p>=0.05): {equivalentes}")
if len(equivalentes) > 1:
    print(f"   -> Sus diferencias NO son distinguibles del ruido de particionado.")
    print(f"3. Criterio de desempate: parsimonia (navaja de Occam).")
    print(f"   -> Se elige el más simple: '{MEJOR_MODELO}' (complejidad {COMPLEJIDAD.get(MEJOR_MODELO,3)}/5).")
else:
    print(f"3. El líder es significativamente superior a todos: no hay desempate.")
c = cmp_tab.loc[MEJOR_MODELO]
print(f"4. Mejor modelo: PR-AUC={c['PR-AUC medio']:.4f} +/- {c['sd']:.4f}")
print(f"   Configuración elegida por la búsqueda (§4.3):")
if MEJOR_MODELO in leaderboard.index:
    print(f"      escalador  = {leaderboard.loc[MEJOR_MODELO,'escalador']}")
    print(f"      muestreo   = {leaderboard.loc[MEJOR_MODELO,'muestreo']}")
    print(f"      hiperparám = {leaderboard.loc[MEJOR_MODELO,'hiperparámetros']}")
base = leaderboard["PR-AUC"].min()
print(f"5. Ganancia vs. el peor del pool: +{S[MEJOR_MODELO].mean()-base:.4f} PR-AUC.")
print("="*70); print(f"MEJOR MODELO -> {MEJOR_MODELO}".center(70)); print("="*70)

plt.figure(figsize=(9, 4.5))
orden = S.mean().sort_values(ascending=False).index
sns.boxplot(data=S[orden], orient="h")
sns.stripplot(data=S[orden], orient="h", color="black", size=3, alpha=.5)
plt.axvline(S[líder].mean(), color="red", ls="--", lw=1, label=f"media del líder")
plt.xlabel("PR-AUC (15 mediciones de CV repetida)")
plt.title("Distribución del desempeño — el solapamiento justifica la prueba estadística")
plt.legend(); plt.tight_layout(); plt.show()
```

> **Fundamento del criterio de selección.** El *boxplot* muestra visualmente lo que la prueba
> formaliza: si las cajas de los finalistas **se solapan ampliamente**, sus diferencias de
> media son ruido. La prueba de Wilcoxon lo convierte en un criterio objetivo, y la
> parsimonia resuelve el empate de forma **defendible ante el profesor**: no se eligió el
> algoritmo de mayor popularidad, sino aquel que la evidencia respalda al menor costo de complejidad.


## 5.2 Umbral operativo

Los modelos entregan **probabilidades**; el umbral 0.5 es una convención arbitraria, no
una decisión óptima. Aquí se elige explícitamente según el objetivo:

```python
proba = modelo_final.predict_proba(X_te[FEATURES])[:, 1]
prec, rec, thr = precision_recall_curve(y_te, proba)
f1c = 2*prec*rec/(prec+rec+1e-9)
thr_f1 = thr[int(np.argmax(f1c[:-1]))]

# umbral por costo: perder un cliente cuesta 5x contactar innecesariamente
COSTO_FN, COSTO_FP = 5.0, 1.0
grid_t = np.linspace(0.05, 0.95, 181)
costos = [COSTO_FN*((y_te==1)&(proba<t)).sum() + COSTO_FP*((y_te==0)&(proba>=t)).sum()
          for t in grid_t]
thr_costo = grid_t[int(np.argmin(costos))]

fig, ax = plt.subplots(1, 2, figsize=(13, 4.2))
ax[0].plot(thr, f1c[:-1]); ax[0].axvline(thr_f1, color="red", ls="--",
    label=f"máx F1 = {thr_f1:.2f}"); ax[0].axvline(0.5, color="gray", ls=":", label="0.5")
ax[0].set_title("F1 vs. umbral"); ax[0].set_xlabel("umbral"); ax[0].legend()
ax[1].plot(grid_t, costos); ax[1].axvline(thr_costo, color="green", ls="--",
    label=f"mín. costo = {thr_costo:.2f}")
ax[1].set_title(f"Costo esperado (FN={COSTO_FN}x, FP={COSTO_FP}x)")
ax[1].set_xlabel("umbral"); ax[1].legend()
plt.tight_layout(); plt.show()

filas = []
for n, t in [("0.50 (convención)",0.5), ("máx. F1",thr_f1), ("mín. costo negocio",thr_costo)]:
    p = (proba >= t).astype(int)
    filas.append({"umbral":n, "valor":round(t,3), "Precision":precision_score(y_te,p),
        "Recall":recall_score(y_te,p), "F1":f1_score(y_te,p),
        "costo":COSTO_FN*((y_te==1)&(p==0)).sum()+COSTO_FP*((y_te==0)&(p==1)).sum()})
display(pd.DataFrame(filas).set_index("umbral").round(4))
print("-> El umbral cambia el resultado de negocio MÁS que cambiar de modelo.")
```

## 5.3 Diagnóstico del modelo seleccionado sobre el conjunto de prueba


```python
UMBRAL_OP = thr_f1     # cambiar a thr_costo si prima el criterio de costo
pred_op = (proba >= UMBRAL_OP).astype(int)

from sklearn.metrics import brier_score_loss
print(f"Modelo: {MEJOR_MODELO} | umbral operativo: {UMBRAL_OP:.3f}\n")
print(classification_report(y_te, pred_op, target_names=["No fuga","Fuga"], digits=3))

# --- las 11 métricas en el TEST, a ambos umbrales ---
def metricas_test(yv, p, sc):
    tn, fp, fn, tp = confusion_matrix(yv, p).ravel()
    return {"PR-AUC":average_precision_score(yv,sc), "ROC-AUC":roc_auc_score(yv,sc),
            "F1":f1_score(yv,p), "F2":fbeta_score(yv,p,beta=2),
            "Precision":precision_score(yv,p), "Recall":recall_score(yv,p),
            "Accuracy":accuracy_score(yv,p), "BalancedAcc":balanced_accuracy_score(yv,p),
            "Especificidad":tn/(tn+fp), "MCC":matthews_corrcoef(yv,p),
            "Kappa":cohen_kappa_score(yv,p), "Brier":brier_score_loss(yv,sc)}
tab_test = pd.DataFrame({
    "umbral 0.50": metricas_test(y_te, (proba>=0.5).astype(int), proba),
    f"umbral operativo {UMBRAL_OP:.2f}": metricas_test(y_te, pred_op, proba),
})
print("=== MÉTRICAS EN EL TEST (nunca usado hasta ahora) ===")
display(tab_test.round(4))

fig, ax = plt.subplots(1, 3, figsize=(16, 4.5))
cm = confusion_matrix(y_te, pred_op)
ConfusionMatrixDisplay(cm, display_labels=["No fuga","Fuga"]).plot(
    cmap="Blues", ax=ax[0], colorbar=False)
ax[0].set_title(f"Matriz de confusión (umbral {UMBRAL_OP:.2f})")
fpr, tpr, _ = roc_curve(y_te, proba)
ax[1].plot(fpr, tpr, label=f"AUC = {roc_auc_score(y_te, proba):.3f}")
ax[1].plot([0,1],[0,1],"--",color="gray"); ax[1].set_title("Curva ROC")
ax[1].set_xlabel("FPR"); ax[1].set_ylabel("TPR"); ax[1].legend()
ax[2].plot(rec, prec, label=f"PR-AUC = {average_precision_score(y_te, proba):.3f}")
ax[2].axhline(y_te.mean(), ls="--", color="gray", label=f"azar = {y_te.mean():.3f}")
ax[2].set_title("Curva Precision-Recall"); ax[2].set_xlabel("Recall")
ax[2].set_ylabel("Precision"); ax[2].legend()
plt.suptitle(f"Diagnóstico del mejor modelo: {MEJOR_MODELO}", y=1.03); plt.tight_layout(); plt.show()

tn, fp, fn, tp = cm.ravel()
print(f"Lectura de negocio (test de {len(y_te)} clientes):")
print(f"   Fugados detectados      : {tp} de {tp+fn} ({tp/(tp+fn)*100:.1f}% de recall)")
print(f"   Fugados NO detectados   : {fn}  <- pérdida real")
print(f"   Falsas alarmas          : {fp}  <- costo de campaña innecesaria")
print(f"   Precisión de la campaña : {tp/(tp+fp)*100:.1f}% de los contactados sí se fugaría")
```

# 6. Aplicación del conocimiento a la decisión de negocio

**Objetivo:** convertir el clasificador en **decisiones de negocio cuantificadas**. Un
modelo con PR-AUC 0.65 no vale nada por sí solo: vale por las acciones que habilita.

Esta fase agrega **herramientas analíticas que el clasificador no puede responder solo**:

| Herramienta | Pregunta que responde | Decisión que habilita |
|---|---|---|
| **Calibración** (§6.1) | ¿Puedo confiar en la probabilidad como probabilidad? | Usar el modelo para calcular valor esperado |
| **Deciles / Lift** (§6.3) | ¿A cuántos clientes contacto? | Tamaño de la campaña |
| **Simulación económica** (§6.4) | ¿Cuánto dinero gano/pierdo? | Umbral **por ROI real**, no por F1 |
| **Matriz Valor × Riesgo** (§6.5) | ¿A **quiénes** priorizo? | Segmentación de cartera |
| **Análisis de supervivencia** (§6.6) | ¿**Cuándo** se fugan? | *Timing* de la intervención |
| **Clustering** (§6.7) | ¿Qué **perfiles** existen? | Propuestas de valor diferenciadas |
| **Contrafactuales** (§6.8) | ¿Qué **palanca** mueve el riesgo? | Priorización de los ejes de trabajo |

## 6.1 Calibración de probabilidades: requisito para la evaluación económica

**Un problema serio que debemos resolver antes de seguir.** El mejor modelo fue elegido por
**PR-AUC**, que mide la calidad del *ranking*. Pero muchos modelos —especialmente los que
usan `class_weight="balanced"` o remuestreo— **distorsionan las probabilidades**: para
ordenar clientes funcionan, pero su "0.7" **no significa 70% de probabilidad real**.

Esto es irrelevante para rankear... y **fatal** para calcular ROI, porque el valor esperado
es `P(fuga) × valor`. Si P está inflada, todo el negocio se calcula mal. Lo medimos:

```python
from sklearn.metrics import brier_score_loss
from sklearn.model_selection import cross_val_predict

# Probabilidades OUT-OF-FOLD para TODA la base (honestas: cada cliente es
# puntuado por un modelo que no lo vio en su entrenamiento)
riesgo_raw = cross_val_predict(clone(modelo_final), X_all[FEATURES], y_all,
    cv=5, method="predict_proba", n_jobs=-1)[:, 1]

calibrado = CalibratedClassifierCV(clone(modelo_final), method="isotonic", cv=5)
riesgo_cal = cross_val_predict(calibrado, X_all[FEATURES], y_all,
    cv=5, method="predict_proba", n_jobs=-1)[:, 1]

print("="*72)
print(f"Tasa de fuga REAL en la base        : {y_all.mean():.4f}")
print(f"Riesgo medio SIN calibrar           : {riesgo_raw.mean():.4f}  "
      f"(Brier={brier_score_loss(y_all, riesgo_raw):.4f})")
print(f"Riesgo medio CALIBRADO              : {riesgo_cal.mean():.4f}  "
      f"(Brier={brier_score_loss(y_all, riesgo_cal):.4f})")
print("-"*72)
print(f"PR-AUC sin calibrar : {average_precision_score(y_all, riesgo_raw):.4f}")
print(f"PR-AUC calibrado    : {average_precision_score(y_all, riesgo_cal):.4f}"
      "   <- el RANKING no cambia")
print("="*72)

fig, ax = plt.subplots(1, 2, figsize=(14, 4.8))
for nombre, r in [("Sin calibrar", riesgo_raw), ("Calibrado (isotónica)", riesgo_cal)]:
    fp, mp = calibration_curve(y_all, r, n_bins=10, strategy="quantile")
    ax[0].plot(mp, fp, "o-", label=nombre)
ax[0].plot([0,1],[0,1],"--",color="gray",label="calibración perfecta")
ax[0].set_xlabel("probabilidad predicha"); ax[0].set_ylabel("fracción real de fuga")
ax[0].set_title("Curva de calibración"); ax[0].legend()
ax[1].hist([riesgo_raw, riesgo_cal], bins=30, label=["Sin calibrar","Calibrado"])
ax[1].axvline(y_all.mean(), color="red", ls="--", label=f"tasa real={y_all.mean():.3f}")
ax[1].set_title("Distribución del riesgo estimado"); ax[1].legend()
plt.tight_layout(); plt.show()

RIESGO = riesgo_cal      # <- de aquí en adelante se decide con probabilidades calibradas
```

> **Decisión §6.1.** Se adopta el modelo **calibrado** para toda la fase de decisión. El
> ranking (y por tanto el PR-AUC y el *targeting*) es idéntico, pero ahora `P=0.7`
> **significa** 70% de probabilidad de fuga. Sin este paso, los cálculos de ROI de §6.4
> estarían inflados y llevarían a **sobre-invertir** en retención.
>
> **Regla general:** calibrar para decidir con dinero; no hace falta para solo ordenar.

## 6.2 Puntuación de la cartera completa

Puntuamos a todos los clientes con probabilidades *out-of-fold* y les asociamos su valor
económico. `ARPU` (*Average Revenue Per User*) = cargo mensual; el **valor anual en riesgo**
es la métrica que traduce el modelo a pesos.

```python
cartera = pd.DataFrame({
    "riesgo":       RIESGO,
    "churn_real":   y_all.values,
    "arpu":         telco["MonthlyCharges"].values,
    "tenure":       telco["tenure"].values,
    "contrato":     telco["Contract"].values,
    "internet":     telco["InternetService"].values,
    "pago":         telco["PaymentMethod"].values,
    "tech_support": telco["TechSupport"].values,
    "seguridad":    telco["OnlineSecurity"].values,
})
cartera["valor_anual"]     = cartera["arpu"] * 12
cartera["ingreso_riesgo"]  = cartera["valor_anual"] * cartera["riesgo"]   # valor esperado

print(f"Clientes: {len(cartera):,}")
print(f"Ingreso anual total de la cartera : ${cartera['valor_anual'].sum():,.0f}")
print(f"Ingreso anual EN RIESGO (esperado): ${cartera['ingreso_riesgo'].sum():,.0f} "
      f"({100*cartera['ingreso_riesgo'].sum()/cartera['valor_anual'].sum():.1f}% del total)")
print(f"ARPU medio: ${cartera['arpu'].mean():.2f} | "
      f"ARPU de los que fugan: ${cartera.loc[cartera.churn_real==1,'arpu'].mean():.2f}")
display(cartera.describe().T.round(3))
```

> **Primer número para la gerencia.** El ingreso anual en riesgo es la cifra que justifica
> el proyecto completo — y ojo con el segundo dato: **los clientes que fugan tienen ARPU
> más alto que el promedio**. No estamos perdiendo clientes marginales, estamos perdiendo
> a los que más facturan. Eso solo, ya cambia la prioridad del negocio.

## 6.3 Deciles de riesgo, *lift* y ganancia acumulada

**Pregunta:** si solo puedo contactar a una fracción de la cartera, ¿a quiénes?
Ordenamos por riesgo y partimos en 10 grupos iguales. El ***lift*** indica cuántas veces
más fuga concentra un decil respecto del azar.

```python
cartera["decil"] = pd.qcut(cartera["riesgo"].rank(method="first"), 10,
                           labels=False).astype(int)
tasa_base = cartera["churn_real"].mean()

dec = (cartera.groupby("decil")
       .agg(clientes=("churn_real","size"), fugados=("churn_real","sum"),
            tasa_fuga=("churn_real","mean"), riesgo_medio=("riesgo","mean"),
            arpu=("arpu","mean"), ingreso_riesgo=("ingreso_riesgo","sum"))
       .sort_index(ascending=False))
dec.index = [f"D{10-i}" for i in range(len(dec))]      # D1 = mayor riesgo
dec["lift"] = dec["tasa_fuga"] / tasa_base
dec["% fugados capturados"] = 100*dec["fugados"].cumsum()/cartera["churn_real"].sum()
dec["% cartera contactada"] = 100*dec["clientes"].cumsum()/len(cartera)
print("=== DECILES DE RIESGO (D1 = mayor riesgo) ===")
display(dec.round(3))

fig, ax = plt.subplots(1, 3, figsize=(17, 4.5))
ax[0].bar(dec.index, dec["lift"], color=sns.color_palette("deep")[0])
ax[0].axhline(1, color="red", ls="--", label="azar (lift=1)")
ax[0].set_title("Lift por decil"); ax[0].set_ylabel("veces sobre el azar"); ax[0].legend()

ax[1].plot(dec["% cartera contactada"], dec["% fugados capturados"], "o-", label="modelo")
ax[1].plot([0,100],[0,100],"--",color="gray",label="azar")
ax[1].set_xlabel("% de cartera contactada"); ax[1].set_ylabel("% de fugados capturados")
ax[1].set_title("Curva de ganancia acumulada"); ax[1].legend(); ax[1].grid(alpha=.3)

ax[2].bar(dec.index, dec["ingreso_riesgo"], color=sns.color_palette("deep")[3])
ax[2].set_title("Ingreso anual en riesgo por decil"); ax[2].set_ylabel("$")
plt.tight_layout(); plt.show()

d3 = dec.iloc[2]["% fugados capturados"]; d2 = dec.iloc[1]["% fugados capturados"]
print(f"Contactando el 20% de mayor riesgo se captura el {d2:.1f}% de los fugados.")
print(f"Contactando el 30% de mayor riesgo se captura el {d3:.1f}% de los fugados.")
print(f"Lift del decil D1: {dec.iloc[0]['lift']:.2f}x sobre el azar.")
```

> **Cómo se lee.** La **curva de ganancia** es la herramienta de *targeting* más usada en
> retención: cuanto más se despegue de la diagonal, mejor discrimina el modelo. Si con el
> **30% de la cartera se captura ~65-70% de los fugados**, la campaña es viable; contactar
> a la totalidad de la cartera resultaría económicamente ineficiente.
>
> El tercer panel agrega la dimensión que falta en el análisis clásico: **no todos los
> deciles valen lo mismo en dinero**. Un decil con alta fuga pero ARPU bajo puede aportar
> menos ingreso en riesgo que uno de riesgo medio con ARPU alto.


## 6.4 Simulación económica: determinación del umbral de máximo beneficio

En §5.2 el umbral se eligió maximizando F1 y con un costo supuesto FN=5×FP. Eso era una
aproximación. Ahora lo hacemos **con la economía real de una campaña de retención**:

$$\text{Beneficio}(u) = \underbrace{N_{TP}(u)\cdot\tau\cdot \text{ARPU}\cdot m\cdot H}_{\text{ingreso salvado}} - \underbrace{N_{contactados}(u)\cdot c}_{\text{costo campaña}} - \underbrace{N_{TP}(u)\cdot\tau\cdot d}_{\text{costo del descuento}}$$

donde `τ` = tasa de éxito de la retención, `m` = margen, `H` = horizonte en meses,
`c` = costo de contactar, `d` = costo del incentivo entregado a quien acepta.

> **Los parámetros son supuestos explícitos y deben validarse con el negocio.** Por eso se
> incluye un **análisis de sensibilidad**: lo importante no es el número exacto, sino
> **cuán robusta es la decisión** ante esos supuestos.

```python
# ---- parámetros económicos (AJUSTAR con el negocio) ----
COSTO_CONTACTO   = 15.0    # $ por contacto (call center, mail, gestión)
TASA_EXITO       = 0.30    # % de los que iban a fugarse y se retienen tras la oferta
MARGEN           = 0.35    # margen sobre ingreso
HORIZONTE_MESES  = 12      # meses de ingreso que se salvan
COSTO_INCENTIVO  = 60.0    # $ del descuento/beneficio entregado al que acepta

def beneficio(umbral, tasa_exito=TASA_EXITO, costo=COSTO_CONTACTO,
              incentivo=COSTO_INCENTIVO, margen=MARGEN, horizonte=HORIZONTE_MESES):
    contactados = cartera["riesgo"] >= umbral
    n_cont = int(contactados.sum())
    if n_cont == 0: return dict(umbral=umbral, n_contactados=0, salvados=0,
                                ingreso=0.0, costo=0.0, beneficio=0.0, roi=0.0)
    tp = contactados & (cartera["churn_real"] == 1)
    salvados = tp.sum() * tasa_exito
    ingreso  = (cartera.loc[tp, "arpu"].sum() * tasa_exito) * margen * horizonte
    costo_t  = n_cont*costo + salvados*incentivo
    return dict(umbral=umbral, n_contactados=n_cont, salvados=round(salvados,1),
                ingreso=ingreso, costo=costo_t, beneficio=ingreso-costo_t,
                roi=(ingreso-costo_t)/costo_t if costo_t > 0 else 0)

umbrales = np.linspace(0.02, 0.95, 94)
econ = pd.DataFrame([beneficio(u) for u in umbrales])
u_opt = econ.loc[econ["beneficio"].idxmax(), "umbral"]
mejor = beneficio(u_opt)

fig, ax = plt.subplots(1, 3, figsize=(17, 4.5))
ax[0].plot(econ["umbral"], econ["beneficio"], lw=2)
ax[0].axvline(u_opt, color="green", ls="--", label=f"óptimo={u_opt:.2f}")
ax[0].axvline(UMBRAL_OP, color="orange", ls=":", label=f"umbral F1={UMBRAL_OP:.2f}")
ax[0].axhline(0, color="red", lw=.8)
ax[0].set_xlabel("umbral"); ax[0].set_ylabel("beneficio $")
ax[0].set_title("Beneficio esperado de la campaña"); ax[0].legend()

ax[1].plot(econ["n_contactados"], econ["beneficio"], lw=2)
ax[1].axhline(0, color="red", lw=.8)
ax[1].set_xlabel("nº de clientes contactados"); ax[1].set_ylabel("beneficio $")
ax[1].set_title("Beneficio vs. tamaño de campaña")

ax[2].plot(econ["umbral"], econ["ingreso"], label="ingreso salvado")
ax[2].plot(econ["umbral"], econ["costo"], label="costo total")
ax[2].axvline(u_opt, color="green", ls="--"); ax[2].set_xlabel("umbral")
ax[2].set_title("Ingreso vs. costo"); ax[2].legend()
plt.tight_layout(); plt.show()

print("="*74)
print("DECISIÓN ECONÓMICA".center(74)); print("="*74)
print(f"Umbral óptimo por ROI      : {u_opt:.3f}")
print(f"Clientes a contactar       : {mejor['n_contactados']:,} "
      f"({100*mejor['n_contactados']/len(cartera):.1f}% de la cartera)")
print(f"Fugas evitadas (esperadas) : {mejor['salvados']:.0f}")
print(f"Ingreso salvado            : ${mejor['ingreso']:,.0f}")
print(f"Costo de la campaña        : ${mejor['costo']:,.0f}")
print(f"BENEFICIO NETO             : ${mejor['beneficio']:,.0f}   (ROI={mejor['roi']*100:.0f}%)")
print("-"*74)
comp = beneficio(UMBRAL_OP)
print(f"Comparación — umbral por F1 ({UMBRAL_OP:.2f}): beneficio ${comp['beneficio']:,.0f} "
      f"con {comp['n_contactados']:,} contactos")
print(f"Ganancia por optimizar con economía real: ${mejor['beneficio']-comp['beneficio']:,.0f}")
print("="*74)
```

### 6.4.1 Análisis de sensibilidad de la decisión

```python
# ¿qué pasa si la tasa de éxito de retención no es 30%?
escenarios = []
for tau in [0.10, 0.20, 0.30, 0.40, 0.50]:
    e = pd.DataFrame([beneficio(u, tasa_exito=tau) for u in umbrales])
    b = e.loc[e["beneficio"].idxmax()]
    escenarios.append({"tasa_éxito":f"{tau:.0%}", "umbral_óptimo":round(b["umbral"],2),
        "contactados":int(b["n_contactados"]), "beneficio":round(b["beneficio"]),
        "ROI":f"{b['roi']*100:.0f}%"})
sens = pd.DataFrame(escenarios).set_index("tasa_éxito")
display(sens)

# punto de equilibrio: tasa de exito minima para no perder plata
taus = np.linspace(0.02, 0.6, 59)
mejor_b = [pd.DataFrame([beneficio(u, tasa_exito=t) for u in umbrales])["beneficio"].max()
           for t in taus]
plt.figure(figsize=(8, 4.2))
plt.plot(taus*100, mejor_b, lw=2); plt.axhline(0, color="red", ls="--", label="equilibrio")
idx = np.argmax(np.array(mejor_b) > 0)
tau_min = taus[idx] if any(np.array(mejor_b) > 0) else np.nan
plt.axvline(tau_min*100, color="green", ls=":", label=f"τ mínima ≈ {tau_min:.0%}")
plt.xlabel("tasa de éxito de retención (%)"); plt.ylabel("beneficio máximo $")
plt.title("Punto de equilibrio de la campaña"); plt.legend(); plt.tight_layout(); plt.show()
print(f"La campaña es rentable solo si la retención supera ≈ {tau_min:.0%} de éxito.")
print("-> Este número es la META que debe comprometer el área comercial.")
```

> **Por qué esto es el corazón de la decisión.** El umbral por F1 optimiza una métrica
> estadística; el umbral económico optimiza **el beneficio neto**. Ambos criterios suelen diferir, y esa diferencia
> es dinero real.
>
> La **sensibilidad** entrega algo más valioso que un número: entrega una **condición de
> viabilidad**. Si la campaña solo es rentable con τ > X%, entonces la pregunta para el
> negocio ya no es "¿el modelo es bueno?" sino "**¿nuestra oferta de retención convierte
> sobre X%?**". Eso es accionable — y si no se sabe, hay que medirlo con un piloto.



## 6.5 Matriz Valor × Riesgo: priorización de la cartera

El riesgo por sí solo no basta: retener a un cliente de \$20/mes no es lo mismo que a uno
de \$100/mes. Cruzamos **probabilidad de fuga** (modelo) con **valor** (ARPU anual) para
segmentar la cartera en cuatro cuadrantes con **estrategias distintas**.

```python
u_riesgo = u_opt                                   # umbral económico de §6.4
u_valor  = cartera["valor_anual"].median()

def cuadrante(r):
    alto_r = r["riesgo"] >= u_riesgo
    alto_v = r["valor_anual"] >= u_valor
    if alto_r and alto_v:   return "A. Alto riesgo / Alto valor"
    if alto_r and not alto_v: return "B. Alto riesgo / Bajo valor"
    if not alto_r and alto_v: return "C. Bajo riesgo / Alto valor"
    return "D. Bajo riesgo / Bajo valor"
cartera["cuadrante"] = cartera.apply(cuadrante, axis=1)

matriz = (cartera.groupby("cuadrante")
    .agg(clientes=("riesgo","size"), tasa_fuga_real=("churn_real","mean"),
         riesgo_medio=("riesgo","mean"), arpu=("arpu","mean"),
         tenure=("tenure","mean"), ingreso_anual=("valor_anual","sum"),
         ingreso_en_riesgo=("ingreso_riesgo","sum")))
matriz["% cartera"] = 100*matriz["clientes"]/len(cartera)
matriz["% ingreso en riesgo"] = 100*matriz["ingreso_en_riesgo"]/cartera["ingreso_riesgo"].sum()
display(matriz.round(2))

fig, ax = plt.subplots(1, 2, figsize=(15, 5.5))
col = {"A. Alto riesgo / Alto valor":"#d62728","B. Alto riesgo / Bajo valor":"#ff7f0e",
       "C. Bajo riesgo / Alto valor":"#2ca02c","D. Bajo riesgo / Bajo valor":"#7f7f7f"}
for q, g in cartera.groupby("cuadrante"):
    ax[0].scatter(g["riesgo"], g["valor_anual"], s=7, alpha=.35, label=q, color=col[q])
ax[0].axvline(u_riesgo, color="black", ls="--"); ax[0].axhline(u_valor, color="black", ls="--")
ax[0].set_xlabel("probabilidad de fuga (calibrada)"); ax[0].set_ylabel("valor anual $")
ax[0].set_title("Cartera: Valor vs. Riesgo"); ax[0].legend(markerscale=2, fontsize=8)
matriz["ingreso_en_riesgo"].sort_values().plot(kind="barh", ax=ax[1],
    color=[col[i] for i in matriz["ingreso_en_riesgo"].sort_values().index])
ax[1].set_title("Ingreso anual en riesgo por cuadrante"); ax[1].set_xlabel("$")
plt.tight_layout(); plt.show()

a = matriz.loc["A. Alto riesgo / Alto valor"]
print(f"CUADRANTE A (prioridad máxima): {a['clientes']:.0f} clientes "
      f"({a['% cartera']:.1f}% de la cartera) concentran "
      f"${a['ingreso_en_riesgo']:,.0f} ({a['% ingreso en riesgo']:.1f}%) del ingreso en riesgo.")
```

> **Criterio de asignación diferenciada.** No se trata a todos igual:
>
> - **A) Alto riesgo / Alto valor** → **retención intensiva**: contacto humano, oferta
>   agresiva. Aquí se justifica gastar. Es el foco del presupuesto.
> - **B) Alto riesgo / Bajo valor** → **retención barata y automatizada** (mail, autoservicio).
>   Si el costo de retener supera el valor del cliente, la no intervención constituye la decisión económicamente correcta.
> - **C) Bajo riesgo / Alto valor** → **no contactar, pero monitorear**: son la base del negocio.
>   El riesgo aquí es *activar* una fuga con una llamada innecesaria.
> - **D) Bajo riesgo / Bajo valor** → **no invertir**; monitoreo pasivo.
>
> Este cuadro convierte un modelo en un **plan de asignación de presupuesto**.

## 6.6 Análisis de supervivencia: momento del abandono

> Esta es la pregunta que **el clasificador no puede responder**. Predice *si* un cliente
> se fuga, no *cuándo*. Y el "cuándo" define el **timing** de la retención.

`tenure` no es una variable cualquiera: es una **duración**. Los clientes que fugaron
tienen `tenure` = tiempo hasta el evento; los que siguen activos están **censurados por la
derecha** (sabemos que sobrevivieron *al menos* ese tiempo). Ese es exactamente el diseño
del estimador **Kaplan-Meier**, que maneja la censura correctamente — cosa que un simple
promedio de `tenure` haría mal.

*(Implementado sin dependencias externas para que corra en cualquier entorno.)*

```python
def kaplan_meier(duraciones, eventos):
    """Estimador Kaplan-Meier. eventos=1 si fugó (evento), 0 si censurado (sigue activo)."""
    d = pd.DataFrame({"t": np.asarray(duraciones), "e": np.asarray(eventos)}).sort_values("t")
    en_riesgo, S, out = len(d), 1.0, []
    for t in np.sort(d["t"].unique()):
        n_t     = int((d["t"] == t).sum())
        eventos_t = int(((d["t"] == t) & (d["e"] == 1)).sum())
        if en_riesgo > 0 and eventos_t > 0:
            S *= (1 - eventos_t/en_riesgo)
        out.append({"t": t, "S": S, "en_riesgo": en_riesgo, "eventos": eventos_t,
                    "hazard": eventos_t/en_riesgo if en_riesgo else 0.0})
        en_riesgo -= n_t
    return pd.DataFrame(out)

def mediana_sup(km):
    b = km[km["S"] <= 0.5]
    return b["t"].iloc[0] if len(b) else np.nan

km_global = kaplan_meier(cartera["tenure"], cartera["churn_real"])

fig, ax = plt.subplots(1, 3, figsize=(17, 4.6))
ax[0].step(km_global["t"], km_global["S"], where="post", lw=2)
ax[0].axhline(0.5, color="red", ls="--", lw=.8)
ax[0].set_xlabel("meses de antigüedad"); ax[0].set_ylabel("S(t) = prob. de seguir activo")
ax[0].set_title("Curva de supervivencia global"); ax[0].grid(alpha=.3)

resumen = []
for c in cartera["contrato"].unique():
    g = cartera[cartera["contrato"] == c]
    km = kaplan_meier(g["tenure"], g["churn_real"])
    ax[1].step(km["t"], km["S"], where="post", lw=2, label=f"{c} (n={len(g)})")
    resumen.append({"Contrato": c, "clientes": len(g),
                    "S(12 meses)": km.loc[km["t"] <= 12, "S"].min(),
                    "S(24 meses)": km.loc[km["t"] <= 24, "S"].min(),
                    "S(48 meses)": km.loc[km["t"] <= 48, "S"].min(),
                    "mediana (meses)": mediana_sup(km)})
ax[1].axhline(0.5, color="red", ls="--", lw=.8)
ax[1].set_xlabel("meses"); ax[1].set_ylabel("S(t)")
ax[1].set_title("Supervivencia por tipo de contrato"); ax[1].legend(); ax[1].grid(alpha=.3)

hz = km_global[km_global["en_riesgo"] >= 30]
ax[2].plot(hz["t"], hz["hazard"].rolling(3, min_periods=1).mean(), lw=2)
ax[2].set_xlabel("meses de antigüedad"); ax[2].set_ylabel("riesgo instantáneo h(t)")
ax[2].set_title("Función de riesgo: ¿cuándo es más peligroso?"); ax[2].grid(alpha=.3)
plt.tight_layout(); plt.show()

print("=== SUPERVIVENCIA POR CONTRATO ===")
display(pd.DataFrame(resumen).set_index("Contrato").round(3))
pico = hz.loc[hz["hazard"].idxmax()]
print(f"\nRiesgo instantáneo máximo en el mes {pico['t']:.0f} (h={pico['hazard']:.3f}).")
print(f"Supervivencia global a 12 meses: {km_global.loc[km_global['t']<=12,'S'].min():.3f}")
```

> **Tres decisiones que salen de aquí y de ningún otro análisis:**
>
> 1. **El *hazard* no es constante.** El riesgo instantáneo se concentra en los **primeros
>    meses** y decae con la antigüedad. Traducción: la retención es un problema de
>    **onboarding**, no de fidelización tardía. Intervenir en el mes 30 es tarde y caro.
> 2. **El contrato cambia la vida del cliente, no solo su probabilidad.** Comparar las
>    medianas de supervivencia por contrato cuantifica *cuántos meses de vida* aporta migrar
>    a anual — un número directamente monetizable (meses × ARPU × margen).
> 3. **Define la ventana de acción.** Si la función de riesgo alcanza su máximo entre los meses 1 y 12, el
>    programa de retención debe activarse ahí, y el modelo debe puntuar a los clientes
>    **nuevos** con prioridad.

## 6.7 Segmentación no supervisada: identificación de perfiles

El modelo supervisado dice *quién* se fuga; el *clustering* dice *cómo son*. Es un lente
complementario: agrupa clientes por similitud **sin mirar la etiqueta**, y luego medimos la
fuga en cada grupo. Si aparecen clusters con fuga muy distinta, tenemos **segmentos
naturales** para propuestas de valor diferenciadas.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

prep_clu = build_prep(NUM_F, CAT_F).fit(X_all[FEATURES])
Z = prep_clu.transform(X_all[FEATURES])
Z = Z.toarray() if hasattr(Z, "toarray") else Z

sil = {}
for k in range(2, 7):
    lab = KMeans(k, n_init=10, random_state=RANDOM_STATE).fit_predict(Z)
    sil[k] = silhouette_score(Z, lab, sample_size=2000, random_state=RANDOM_STATE)
K = max(sil, key=sil.get)
print("Silueta por k:", {k: round(v,3) for k,v in sil.items()}, "-> k elegido =", K)

cartera["cluster"] = KMeans(K, n_init=10, random_state=RANDOM_STATE).fit_predict(Z)
perfil = (cartera.groupby("cluster")
    .agg(clientes=("riesgo","size"), tasa_fuga=("churn_real","mean"),
         riesgo_medio=("riesgo","mean"), arpu=("arpu","mean"),
         tenure=("tenure","mean"), ingreso_en_riesgo=("ingreso_riesgo","sum")))
perfil["% cartera"] = 100*perfil["clientes"]/len(cartera)
perfil["lift_fuga"] = perfil["tasa_fuga"]/cartera["churn_real"].mean()
display(perfil.round(3).sort_values("tasa_fuga", ascending=False))

print("\n=== COMPOSICIÓN DE CADA CLUSTER ===")
for c in sorted(cartera["cluster"].unique()):
    g = cartera[cartera["cluster"] == c]
    print(f"\nCluster {c}: {len(g)} clientes | fuga={g['churn_real'].mean():.1%} | "
          f"ARPU=${g['arpu'].mean():.0f} | tenure={g['tenure'].mean():.0f} meses")
    print(f"   contrato : {g['contrato'].value_counts(normalize=True).round(2).to_dict()}")
    print(f"   internet : {g['internet'].value_counts(normalize=True).round(2).to_dict()}")

fig, ax = plt.subplots(1, 2, figsize=(14, 4.5))
perfil.sort_values("tasa_fuga")[["tasa_fuga","riesgo_medio"]].plot(kind="barh", ax=ax[0])
ax[0].axvline(cartera["churn_real"].mean(), color="red", ls="--", label="tasa global")
ax[0].set_title("Fuga por cluster"); ax[0].legend()
ax[1].scatter(perfil["arpu"], perfil["tasa_fuga"], s=perfil["clientes"]/4,
              c=range(len(perfil)), cmap="tab10")
for i, r in perfil.iterrows(): ax[1].annotate(f"C{i}", (r["arpu"], r["tasa_fuga"]))
ax[1].set_xlabel("ARPU medio $"); ax[1].set_ylabel("tasa de fuga")
ax[1].set_title("Perfiles: valor vs. fuga (tamaño = nº clientes)")
plt.tight_layout(); plt.show()
```

> **Lectura.** La **silueta** suele ser baja (0.1–0.25) en datos con muchas variables
> categóricas: los clusters no están nítidamente separados. **Eso no los invalida** como
> herramienta de gestión, pero obliga a honestidad: son **agrupaciones útiles, no
> fronteras naturales**. Lo relevante es si el `lift_fuga` entre clusters es
> suficientemente distinto como para justificar tratamientos diferenciados.

## 6.8 Simulación contrafactual: efecto potencial de cada atributo

Usamos el modelo como **simulador**: tomamos a los clientes de un segmento, les cambiamos
**una sola característica** y volvemos a predecir. La diferencia de riesgo estima el efecto
de esa palanca — y así **priorizamos los ejes de trabajo con números**.

> ### ⚠️ Advertencia causal — leer antes de usar estos números
>
> **Esto NO es un efecto causal.** El modelo aprendió **correlaciones** de datos
> observacionales. Que los clientes con contrato anual fuguen menos **no prueba** que
> migrar a un cliente a contrato anual reduzca su fuga: puede ser que los clientes ya
> comprometidos *elijan* el contrato anual (**causalidad inversa**) o que un tercer factor
> explique ambos (**confusión**).
>
> Estos números son un **ranking de hipótesis priorizadas**, no una promesa de resultados.
> La única forma de establecer causalidad es un **experimento controlado (A/B test)** —
> cuyo diseño se propone en §6.9. Presentarlos como efectos causales sería un error grave.

```python
def simular(cambios: dict, mask=None):
    """Aplica `cambios` a los clientes de `mask`, recalcula features y re-predice."""
    Xm = X_raw.copy()
    idx = Xm.index if mask is None else Xm.index[mask]
    for col, val in cambios.items():
        Xm.loc[idx, col] = val
    return calib_full.predict_proba(fe_telco(Xm)[FEATURES])[:, 1]

# modelo calibrado ajustado a toda la base, para simular
calib_full = CalibratedClassifierCV(clone(modelo_final), method="isotonic",
                                    cv=5).fit(X_all[FEATURES], y_all)
riesgo_base = calib_full.predict_proba(X_all[FEATURES])[:, 1]

# palancas: (nombre, cambios, sobre qué clientes aplica)
es_m2m   = (X_raw["Contract"] == "Month-to-month").values
sin_ts   = (X_raw["TechSupport"] == "No").values
sin_sec  = (X_raw["OnlineSecurity"] == "No").values
es_check = X_raw["PaymentMethod"].str.contains("Electronic check").values
es_fibra = (X_raw["InternetService"] == "Fiber optic").values

PALANCAS = [
 ("Migrar a contrato ANUAL",        {"Contract":"One year"},                      es_m2m),
 ("Migrar a contrato BIANUAL",      {"Contract":"Two year"},                      es_m2m),
 ("Activar TechSupport",            {"TechSupport":"Yes"},                        sin_ts),
 ("Activar OnlineSecurity",         {"OnlineSecurity":"Yes"},                     sin_sec),
 ("Activar Support + Security",     {"TechSupport":"Yes","OnlineSecurity":"Yes"}, sin_ts & sin_sec),
 ("Migrar a pago automático",       {"PaymentMethod":"Bank transfer (automatic)"},es_check),
 ("Fibra: activar seguridad",       {"OnlineSecurity":"Yes","TechSupport":"Yes"}, es_fibra & sin_sec),
]

filas = []
for nombre, cambios, mask in PALANCAS:
    if mask.sum() == 0: continue
    r_new = simular(cambios, mask)
    r0, r1 = riesgo_base[mask].mean(), r_new[mask].mean()
    arpu_seg = cartera.loc[mask, "arpu"].sum()
    filas.append({
        "Palanca": nombre, "clientes afectados": int(mask.sum()),
        "riesgo antes": r0, "riesgo después": r1, "Δ riesgo": r1-r0,
        "reducción %": 100*(r1-r0)/r0 if r0 else 0,
        "fugas evitadas (est.)": (r0-r1)*mask.sum(),
        "ingreso anual protegido": (r0-r1)*arpu_seg*12})
palancas = pd.DataFrame(filas).set_index("Palanca").sort_values("Δ riesgo")
display(palancas.round(3))

fig, ax = plt.subplots(1, 2, figsize=(15, 5))
palancas["Δ riesgo"].plot(kind="barh", ax=ax[0],
    color=["#2ca02c" if v < 0 else "#d62728" for v in palancas["Δ riesgo"]])
ax[0].axvline(0, color="black", lw=.8)
ax[0].set_title("Cambio de riesgo por palanca (negativo = mejora)")
palancas["ingreso anual protegido"].sort_values().plot(kind="barh", ax=ax[1],
    color=sns.color_palette("deep")[2])
ax[1].set_title("Ingreso anual protegido estimado $")
plt.tight_layout(); plt.show()
print("Recordatorio: son ASOCIACIONES priorizadas, no efectos causales (ver §6.9).")
```

## 6.9 Validación causal: diseño experimental requerido

Los contrafactuales de §6.8 **ordenan hipótesis**; no las prueban. Para convertir la palanca
principal en una decisión con respaldo causal, se propone un **piloto A/B**:

| Elemento | Definición |
|---|---|
| **Hipótesis** | Ofrecer migración a contrato anual reduce la fuga a 12 meses en el segmento de alto riesgo |
| **Población** | Cuadrante A (§6.5): alto riesgo × alto valor, contrato *month-to-month* |
| **Asignación** | Aleatoria: 50% recibe la oferta (tratamiento), 50% gestión habitual (control) |
| **Métrica primaria** | Tasa de fuga a 12 meses |
| **Métricas secundarias** | Tasa de aceptación, ARPU post-oferta, margen neto |
| **Análisis** | Diferencia de proporciones + intervalo de confianza; *uplift* = fuga(control) − fuga(tratamiento) |

> **Por qué importa.** Sin aleatorización, ofrecer retención solo a los de alto riesgo y
> luego comparar contra los de bajo riesgo **confundiría** el efecto de la oferta con el
> efecto del riesgo. Además, existe el riesgo de activación de clientes latentes: usuarios que no
> pensaban irse y a los que la llamada les *recuerda* que pueden hacerlo. El grupo control
> es la única forma de detectarlo.
>
> El tamaño muestral se puede estimar con el efecto esperado de §6.8 como hipótesis inicial:
> ese es el aporte del modelo al diseño experimental.

```python
# tamaño muestral requerido para detectar el efecto (potencia estadística)
from statistics import NormalDist
def n_por_grupo(p_control, efecto_abs, alpha=0.05, potencia=0.80):
    z_a = NormalDist().inv_cdf(1-alpha/2); z_b = NormalDist().inv_cdf(potencia)
    p2 = p_control - efecto_abs
    p_bar = (p_control+p2)/2
    return int(np.ceil(2*p_bar*(1-p_bar)*(z_a+z_b)**2 / efecto_abs**2))

q_a = cartera[cartera["cuadrante"] == "A. Alto riesgo / Alto valor"]
p_ctrl = q_a["churn_real"].mean()
print(f"Tasa de fuga observada en el cuadrante A: {p_ctrl:.1%}")
print(f"Clientes disponibles en el cuadrante A  : {len(q_a):,}\n")
print("Tamaño muestral POR GRUPO según el efecto que se quiera detectar:")
for ef in [0.03, 0.05, 0.08, 0.10]:
    n = n_por_grupo(p_ctrl, ef)
    viable = "viable" if 2*n <= len(q_a) else "NO alcanza la población"
    print(f"   reducir la fuga en {ef:.0%} -> n={n:,} por grupo ({2*n:,} total) -> {viable}")
```

## 7.1 Respuesta a las preguntas de investigación

### P1. ¿Sobre qué conjunto de datos es posible construir un modelo confiable?

Sobre el conjunto **IBM Telco Customer Churn** (7.043 clientes, 21 variables). El candidato alternativo fue descartado tras aplicar tres pruebas de aptitud: un árbol de decisión de profundidad 3 alcanzó un AUC de 0,932, lo que evidencia una etiqueta generada de forma determinista; la validación adversarial arrojó un AUC de 0,743, incompatible con la hipótesis de distribuciones equivalentes entre train y test; y los modelos entrenados sobre él colapsaron, prediciendo abandono en el 86,5 % del test cuando la tasa real era de 47,4 %.

La conclusión metodológica asociada es que **un F1 de 0,999 no constituyó un logro sino una señal de alerta**. La auditoría previa al modelamiento evitó reportar como exitoso un resultado que era un artefacto del proceso de generación de los datos.

### P2. ¿Qué algoritmo ofrece el mejor desempeño y con qué criterio se selecciona?

El modelo con mejor desempeño fue **Gradient Boosting** (PR-AUC = 0,6641 ± 0,0181), configurado con SMOTE a razón 0,5, profundidad 2, tasa de aprendizaje 0,05 y 300 estimadores. No obstante, el hallazgo relevante es el contexto de esa elección:

- Los diez mejores algoritmos se ubicaron dentro de un rango de **0,013 de PR-AUC**. Una regresión logística adecuadamente ajustada (0,6587) resultó equivalente a XGBoost (0,6596).
- El criterio de selección no fue la media observada. La prueba de **Wilcoxon pareada** sobre quince mediciones indicó que la diferencia entre el líder (*Stacking*, 0,6646) y Gradient Boosting (0,6641) no es distinguible del ruido de particionado (*p* = 0,303). Ante la equivalencia estadística de tres modelos, se aplicó el **criterio de parsimonia**.
- Respecto del balanceo de clases, once estrategias evaluadas desplazaron el *recall* entre 0,52 y 0,84 modificando el PR-AUC en apenas 0,013. El remuestreo **reasigna el punto de operación sin incorporar información**. Se verificó además que su aplicación fuera del procedimiento de validación produce una inflación artificial del 28 % del desempeño estimado.

### P3. ¿A qué clientes debe dirigirse la retención y qué ingreso está expuesto?

La cartera representa un ingreso anual de **US$ 5,47 millones**, de los cuales **US$ 1,66 millones (30,4 %)** se encuentran expuestos al abandono. La distribución de dicha exposición no es homogénea:

- El **cuadrante A** (alto riesgo y alto valor) agrupa **2.245 clientes —el 31,9 % de la cartera— y concentra US$ 1,20 millones, equivalentes al 72,4 % del ingreso expuesto**. Sobre este segmento debe concentrarse el presupuesto de retención.
- El **cuadrante B** (alto riesgo, bajo valor) admite mecanismos automatizados de bajo costo. Cuando el costo de retención supera el valor del cliente, la no intervención constituye la decisión económicamente correcta.
- El **cuadrante C** (bajo riesgo, alto valor) debe **excluirse del contacto**. Constituye la base de ingresos consolidada y su intervención introduce el riesgo de activar clientes latentes.

Un hallazgo adicional refuerza la prioridad: los clientes que abandonan presentan un **ARPU de US$ 74,4 frente a US$ 64,8 del promedio**. La organización no pierde clientes marginales, sino aquellos de mayor facturación. La segmentación no supervisada, aplicada sin utilizar la variable objetivo, identificó de forma independiente el mismo segmento crítico (3.078 clientes, 98 % con contrato mes a mes), lo que respalda la validez del diagnóstico.

### P4. ¿A cuántos clientes contactar, con qué retorno y bajo qué condiciones?

Una campaña dirigida a **3.478 clientes (49,4 % de la cartera)**, definida por el umbral de máximo beneficio (0,19), permitiría evitar **490 abandonos** y generar un beneficio neto de **US$ 74.596**, equivalente a un **ROI de 91 %**. La curva de ganancia acumulada muestra que el 30 % de la cartera de mayor riesgo concentra el 67 % de los abandonos, por lo que el contacto masivo resulta económicamente ineficiente.

Dos precisiones resultan indispensables:

1. **El umbral económico difiere del umbral estadístico.** El umbral que maximiza F1 (0,30) y el que maximiza el beneficio (0,19) no coinciden. La elección del punto de operación es una decisión de negocio y no estadística.
2. **La viabilidad está condicionada.** El análisis de sensibilidad establece que la campaña resulta rentable únicamente si la oferta de retención **convierte por sobre el 6 %**. Este umbral no lo determina el modelo: constituye el compromiso que debe asumir el área comercial, y su medición es la primera etapa de la hoja de ruta (§6.12).

### P5. ¿Cuándo intervenir y qué atributos presentan mayor efecto?

El análisis de supervivencia responde una pregunta que el clasificador no permite abordar. La función de riesgo **alcanza su máximo en el primer mes** y decrece con la antigüedad: el abandono es un fenómeno temprano y, por tanto, la retención constituye un problema de acompañamiento inicial y no de fidelización tardía. La mediana de vida en contrato mes a mes es de **35 meses**, mientras que el 99,6 % de los clientes con contrato bianual permanece activo a los cuatro años. La migración contractual no modifica únicamente una probabilidad: modifica la vida esperada del cliente.

La simulación contrafactual ordena los atributos según su efecto potencial: la migración a contrato bianual reduce el riesgo estimado en **59,9 %** (US$ 786.622 de ingreso protegido), seguida por la activación de servicios de seguridad en clientes de fibra óptica (43,5 %) y la migración a contrato anual (42,3 %).

### P6. ¿Cómo se priorizan las acciones y cómo se validan?

Se definieron **seis ejes de trabajo**, cada uno dimensionado en número de clientes, tasa de abandono e ingreso expuesto (§6.10 y §6.11). La priorización resulta de la relación entre impacto y esfuerzo, y la secuencia recomendada (§6.12) no comienza por el eje de mayor magnitud sino por el de menor esfuerzo, dado que su función es **medir el supuesto crítico** del modelo económico.

Respecto de la validación, los efectos estimados en §6.8 corresponden a **asociaciones y no a relaciones causales**. El modelo fue entrenado sobre datos observacionales, por lo que la menor tasa de abandono de los clientes con contrato anual no permite concluir que migrar a un cliente reduzca su probabilidad de abandono: la relación podría explicarse por causalidad inversa. El mecanismo de validación propuesto es un **experimento aleatorizado** sobre el cuadrante A, cuyo diseño y potencia estadística se especifican en §6.9.

## 7.2 Recomendaciones para el negocio

Sobre la base de la evidencia expuesta, se formulan cuatro recomendaciones:

1. **Concentrar el presupuesto de retención en el cuadrante A.** Un tercio de la cartera concentra casi tres cuartos del ingreso expuesto. De forma simétrica, se recomienda **excluir del contacto al cuadrante C**, cuyo riesgo de abandono es bajo y cuya intervención puede resultar contraproducente.

2. **Definir el volumen de contacto según el umbral económico y no según métricas estadísticas.** La diferencia entre ambos criterios se traduce en beneficio efectivo.

3. **Activar el programa de retención durante el primer año del ciclo de vida**, con énfasis en los primeros meses, que es donde se concentra la función de riesgo. La intervención tardía es menos efectiva y más costosa.

4. **Iniciar la implementación por el eje de menor esfuerzo**, cuyo propósito es estimar empíricamente la tasa de conversión, y validar la palanca contractual mediante un experimento aleatorizado antes de escalarla.

## 7.3 Limitaciones del estudio

1. **El supuesto económico central no se encuentra medido.** La estimación de beneficio depende de la tasa de conversión de la oferta de retención, cuyo valor fue supuesto en 30 %. El análisis de sensibilidad acota el rango de decisión, pero no sustituye su medición empírica.

2. **Ningún efecto se encuentra probado causalmente.** Los resultados de la simulación contrafactual constituyen hipótesis priorizadas mediante evidencia, no estimaciones de efecto causal.

3. **Los datos son de naturaleza transversal.** La ausencia de un componente temporal impide validar el modelo mediante una partición cronológica y anticipar la deriva de concepto, lo que constituye la principal limitación para su operación sostenida.

4. **El modelo identifica manifestaciones y no causas.** Que el segmento de fibra óptica sin servicios de seguridad concentre una mayor tasa de abandono no permite establecer su origen, que podría corresponder a factores de precio, calidad técnica o expectativas. Su diagnóstico requiere investigación complementaria que los datos disponibles no contienen.

5. **La búsqueda de hiperparámetros es incompleta por diseño.** El procedimiento explora un subconjunto del espacio de configuraciones; el presupuesto empleado otorga una probabilidad promedio del 72 % de alcanzar el 5 % superior de las configuraciones por modelo.

## 7.4 Trabajo futuro

Se recomienda incorporar información longitudinal del comportamiento de los clientes que permita validar el modelo mediante particiones temporales y detectar deriva de concepto; ejecutar el diseño experimental propuesto en §6.9 con el fin de estimar efectos causales; y complementar el análisis cuantitativo con investigación cualitativa sobre los segmentos de mayor riesgo, particularmente el de fibra óptica, cuyo comportamiento el modelo describe pero no explica.

## 7.5 Consideración final

El estudio confirma que la aplicación de una metodología sistemática como KDD, complementada con una fase de aplicación del conocimiento, permite desarrollar soluciones de aprendizaje supervisado que combinan rigor metodológico con utilidad práctica. Los dos hallazgos de mayor valor del trabajo no provinieron de la etapa de modelamiento: el primero fue determinar que un conjunto de datos aparentemente óptimo no era apto para el objetivo planteado; el segundo, que la elección del algoritmo resultó irrelevante frente a decisiones como el punto de operación, la segmentación de la cartera y el momento de intervención.

En síntesis, **el valor de un modelo predictivo no reside en sus métricas de clasificación, sino en las decisiones que permite fundamentar**.
